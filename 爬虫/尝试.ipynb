```python
from pymongo import MongoClient,errors
```

```python
client = MongoClient()
client
```

```python
databese=client["xueqiu"]
databese
```

```python
sheet=databese["comments"]
sheet
```

```python
for comment in list_data:
    sub_text = re.sub(u"<.*?>|\\$.*?\\$", "", comment['text'])
    print(sub_text.strip())
    try:
        # sheet.insert_one里面用_id表示key
        sheet.insert_one({'_id':comment['id'],'user_id':comment['user_id'],'time':comment['created_at'],'text':sub_text.strip(),'raw':comment})
        print(comment['id'],'评论插入成功')
    except errors.DuplicateKeyError as e:
        print(comment['id'],'已经存在于数据库')
        pass
```
