```python
# %load data.csv
_id,user_id,time,text,raw
142933428,6690948234,1583334234000,水军们举报还是满厉害的,"{'id': 142933428, 'user_id': 6690948234, 'title': '', 'text': '<a href=""http://xueqiu.com/S/01810"" target=""_blank"">$小米集团-W(<span class=\'highlight\'>01810</span>)$</a> 水军们举报还是满厉害的<img src=""//assets.imedao.com/ugc/images/face/emoji_07_wonderful.png"" title=""[牛]"" alt=""[牛]"" height=""24"" /><img src=""//assets.imedao.com/ugc/images/face/emoji_07_wonderful.png"" title=""[牛]"" alt=""[牛]"" height=""24"" /><br/><img class=""ke_img"" src=""https://xqimg.imedao.com/170a6131489f84e3fe526b8f.jpeg!custom.jpg"" /><br/><img class=""ke_img"" src=""https://xqimg.imedao.com/170a6131489f94e3fcb83a2b.jpeg!custom.jpg"" />', 'created_at': 1583334234000, 'retweet_count': 0, 'reply_count': 0, 'fav_count': 0, 'truncated': False, 'commentId': 0, 'retweet_status_id': 0, 'symbol_id': None, 'description': "" $小米集团-W( <span class='highlight'>01810</span> )$ 水军们举报还是满厉害的 "", 'type': '0', 'source_link': 'http://xueqiu.com/about/mobile-xueqiu', 'edited_at': None, 'pic': 'https://xqimg.imedao.com/170a6131489f84e3fe526b8f.jpeg!thumb.jpg,https://xqimg.imedao.com/170a6131489f94e3fcb83a2b.jpeg!thumb.jpg', 'user': {'subscribeable': False, 'remark': None, 'status': 2, 'common_count': 0, 'domain': None, 'description': None, 'type': '1', 'location': None, 'id': 6690948234, 'stats': None, 'url': None, 'screen_name': '半坡村之流年岁月', 'following': False, 'profile': '/6690948234', 'verified': False, 'blocking': False, 'donate_count': 0, 'friends_count': 37, 'followers_count': 7, 'st_color': '1', 'follow_me': False, 'verified_description': '', 'province': None, 'city': None, 'gender': None, 'step': 'null', 'blog_description': None, 'verified_type': 0, 'last_status_id': 142933534, 'status_count': 248, 'recommend': None, 'intro': None, 'stock_status_count': None, 'allow_all_stock': False, 'stocks_count': None, 'name': None, 'group_ids': None, 'name_pinyin': None, 'screenname_pinyin': None, 'block_status': 0, '_source': None, 'verified_realname': False, 'verified_infos': [{'verified_type': '5', 'verified_desc': '用户已完成实名身份认证'}], 'profile_image_url': 'community/20145/1402406920545-1402406920838.jpg,community/20145/1402406920545-1402406920838.jpg!180x180.png,community/20145/1402406920545-1402406920838.jpg!50x50.png,community/20145/1402406920545-1402406920838.jpg!30x30.png', 'photo_domain': '//xavatar.imedao.com/'}, 'retweeted_status': None, 'target': '/6690948234/142933428', 'fragment': None, 'blocked': False, 'blocking': False, 'topic_pic': None, 'topic_symbol': None, 'topic_title': None, 'topic_desc': None, 'donate_count': 0, 'donate_snowcoin': 0, 'view_count': 0, 'offer': None, 'mark': 0, 'comment': None, 'expend': False, 'rqtype': None, 'rqid': 0, 'firstImg': 'https://xqimg.imedao.com/170a6131489f84e3fe526b8f.jpeg!thumb.jpg', 'card': None, 'favorited': False, 'favorited_created_at': None, 'topic_pic_thumbnail': None, 'topic_pic_headOrPad': None, 'topic_pic_thumbnail_small': None, 'canEdit': True, 'shortStatusMultiPic': True, 'timeBefore': '4分钟前', 'mark_desc': None, 'reward_count': 0, 'reward_user_count': 0, 'paid_mention': None, 'is_answer': False, 'is_refused': False, 'is_reward': True, '_source': 'sc_1:1:01810', 'pic_sizes': [{'width': 1080, 'height': 2340, 'ratio': '0.46', 'score': 1, 'setScore': True, 'setWidth': True, 'setHeight': True, 'setRatio': True}, {'width': 1080, 'height': 2340, 'ratio': '0.46', 'score': 1, 'setScore': True, 'setWidth': True, 'setHeight': True, 'setRatio': True}], 'like_count': 1, 'liked': False, 'promotion_pic': None, 'promotion_url': None, 'promotion_id': 0, 'quote_cards': None, 'is_ss_multi_pic': True, 'excellent_comments': None, 'is_bonus': False, 'source': 'Android'}"
142932876,6690948234,1583333861000,我眼中小米是1/10的apple加 1/2严选 加 1/4飞利浦照明 加 1/5戴森小家电 加 1/5无印良品 加 1/6格力 1/2三星电视 1/7都美的冰洗 加百分百领先的lot 加1/10的阿里云 加1/10京东物流  加1/10的Costa 加1/20的支付宝  唯一比较欠缺的腾讯的社交,"{'id': 142932876, 'user_id': 6690948234, 'title': '', 'text': '<a href=""http://xueqiu.com/S/01810"" target=""_blank"">$小米集团-W(<span class=\'highlight\'>01810</span>)$</a> 我眼中小米是1/10的apple加 1/2严选 加 1/4飞利浦照明 加 1/5戴森小家电 加 1/5无印良品 加 1/6格力 1/2三星电视 1/7都美的冰洗 加百分百领先的lot 加1/10的阿里云 加1/10京东物流  加1/10的Costa 加1/20的支付宝  唯一比较欠缺的腾讯的社交', 'created_at': 1583333861000, 'retweet_count': 1, 'reply_count': 1, 'fav_count': 0, 'truncated': False, 'commentId': 0, 'retweet_status_id': 0, 'symbol_id': None, 'description': "" $小米集团-W( <span class='highlight'>01810</span> )$ 我眼中小米是1/10的apple加 1/2严选 加 1/4飞利浦照明 加 1/5戴森小家电 加 1/5无印良品 加 1/6格力 1/2三星电视 1/7都美的冰洗 加百分百领先的lot 加1/10的阿里云 加1/10京东物流 加1/10的Costa 加1/20的支付宝 唯一比较欠缺的腾讯的社交"", 'type': '0', 'source_link': 'http://xueqiu.com/about/mobile-xueqiu', 'edited_at': None, 'pic': '', 'user': {'subscribeable': False, 'remark': None, 'status': 2, 'common_count': 0, 'domain': None, 'description': None, 'type': '1', 'location': None, 'id': 6690948234, 'stats': None, 'url': None, 'screen_name': '半坡村之流年岁月', 'following': False, 'profile': '/6690948234', 'verified': False, 'blocking': False, 'donate_count': 0, 'friends_count': 37, 'followers_count': 7, 'st_color': '1', 'follow_me': False, 'verified_description': '', 'province': None, 'city': None, 'gender': None, 'step': 'null', 'blog_description': None, 'verified_type': 0, 'last_status_id': 142933534, 'status_count': 248, 'recommend': None, 'intro': None, 'stock_status_count': None, 'allow_all_stock': False, 'stocks_count': None, 'name': None, 'group_ids': None, 'name_pinyin': None, 'screenname_pinyin': None, 'block_status': 0, '_source': None, 'verified_realname': False, 'verified_infos': [{'verified_type': '5', 'verified_desc': '用户已完成实名身份认证'}], 'profile_image_url': 'community/20145/1402406920545-1402406920838.jpg,community/20145/1402406920545-1402406920838.jpg!180x180.png,community/20145/1402406920545-1402406920838.jpg!50x50.png,community/20145/1402406920545-1402406920838.jpg!30x30.png', 'photo_domain': '//xavatar.imedao.com/'}, 'retweeted_status': None, 'target': '/6690948234/142932876', 'fragment': None, 'blocked': False, 'blocking': False, 'topic_pic': None, 'topic_symbol': None, 'topic_title': None, 'topic_desc': None, 'donate_count': 0, 'donate_snowcoin': 0, 'view_count': 0, 'offer': None, 'mark': 0, 'comment': None, 'expend': False, 'rqtype': None, 'rqid': 0, 'firstImg': None, 'card': None, 'favorited': False, 'favorited_created_at': None, 'topic_pic_thumbnail': None, 'topic_pic_headOrPad': None, 'topic_pic_thumbnail_small': None, 'canEdit': True, 'shortStatusMultiPic': False, 'timeBefore': '10分钟前', 'mark_desc': None, 'reward_count': 0, 'reward_user_count': 0, 'paid_mention': None, 'is_answer': False, 'is_refused': False, 'is_reward': True, '_source': 'sc_1:2:01810', 'pic_sizes': [], 'like_count': 1, 'liked': False, 'promotion_pic': None, 'promotion_url': None, 'promotion_id': 0, 'quote_cards': None, 'is_ss_multi_pic': False, 'excellent_comments': None, 'is_bonus': False, 'source': 'Android'}"
142932677,1944137583,1583333726000,首个高端机型小米10系列京东销售数据不错，但是好评率都不高，无论京东还是小米商城，这是什么原因？,"{'id': 142932677, 'user_id': 1944137583, 'title': '', 'text': '首个高端机型小米10系列京东销售数据不错，但是好评率都不高，无论京东还是小米商城，这是什么原因？ <a href=""http://xueqiu.com/S/01810"" target=""_blank"">$小米集团-W(<span class=\'highlight\'>01810</span>)$</a>', 'created_at': 1583333726000, 'retweet_count': 0, 'reply_count': 2, 'fav_count': 0, 'truncated': False, 'commentId': 0, 'retweet_status_id': 0, 'symbol_id': None, 'description': ""首个高端机型小米10系列京东销售数据不错，但是好评率都不高，无论京东还是小米商城，这是什么原因？ $小米集团-W( <span class='highlight'>01810</span> )$ "", 'type': '0', 'source_link': 'http://xueqiu.com/about/mobile-xueqiu', 'edited_at': None, 'pic': '', 'user': {'subscribeable': False, 'remark': None, 'status': 2, 'common_count': 0, 'domain': None, 'description': '', 'type': '1', 'location': None, 'id': 1944137583, 'stats': None, 'url': None, 'screen_name': '而今思惘', 'following': False, 'profile': '/1944137583', 'verified': False, 'blocking': False, 'donate_count': 0, 'friends_count': 113, 'followers_count': 7, 'st_color': '1', 'follow_me': False, 'verified_description': None, 'province': '', 'city': '', 'gender': 'n', 'step': 'null', 'blog_description': None, 'verified_type': 0, 'last_status_id': 142932677, 'status_count': 44, 'recommend': None, 'intro': None, 'stock_status_count': None, 'allow_all_stock': False, 'stocks_count': None, 'name': None, 'group_ids': None, 'name_pinyin': None, 'screenname_pinyin': None, 'block_status': 0, '_source': None, 'verified_realname': False, 'verified_infos': None, 'profile_image_url': 'community/20190/1546520387679-1546520387941.jpg,community/20190/1546520387679-1546520387941.jpg!180x180.png,community/20190/1546520387679-1546520387941.jpg!50x50.png,community/20190/1546520387679-1546520387941.jpg!30x30.png', 'photo_domain': '//xavatar.imedao.com/'}, 'retweeted_status': None, 'target': '/1944137583/142932677', 'fragment': None, 'blocked': False, 'blocking': False, 'topic_pic': None, 'topic_symbol': None, 'topic_title': None, 'topic_desc': None, 'donate_count': 0, 'donate_snowcoin': 0, 'view_count': 0, 'offer': None, 'mark': 0, 'comment': None, 'expend': False, 'rqtype': None, 'rqid': 0, 'firstImg': None, 'card': None, 'favorited': False, 'favorited_created_at': None, 'topic_pic_thumbnail': None, 'topic_pic_headOrPad': None, 'topic_pic_thumbnail_small': None, 'canEdit': True, 'shortStatusMultiPic': False, 'timeBefore': '12分钟前', 'mark_desc': None, 'reward_count': 0, 'reward_user_count': 0, 'paid_mention': None, 'is_answer': False, 'is_refused': False, 'is_reward': True, '_source': 'sc_1:3:01810', 'pic_sizes': [], 'like_count': 0, 'liked': False, 'promotion_pic': None, 'promotion_url': None, 'promotion_id': 0, 'quote_cards': None, 'is_ss_multi_pic': False, 'excellent_comments': None, 'is_bonus': False, 'source': 'iPhone'}"
142932571,4706271602,1583333670000,小米的宣传总是直男风十足。明明一个非常好的“人像留色”功能和卖点，被拍出了使用说明书教程范儿。土！！！如果是华为苹果三星罗永浩来拍，效果不要太骚 网页链接,"{'id': 142932571, 'user_id': 4706271602, 'title': '', 'text': '小米的宣传总是直男风十足。明明一个非常好的“人像留色”功能和卖点，被拍出了使用说明书教程范儿。土！！！<br/><br/>如果是华为苹果三星罗永浩来拍，效果不要太骚<a href=""http://xueqiu.com/S/01810"" target=""_blank"">$小米集团-W(<span class=\'highlight\'>01810</span>)$</a> <a href=""https://weibointl.api.weibo.cn/share/128940778.html?weibo_id=4478884982656036"" title=""https://weibointl.api.weibo.cn/share/128940778.html?weibo_id=4478884982656036"" target=""_blank"" class=""status-link"">网页链接</a><br/><img class=""ke_img"" src=""https://xqimg.imedao.com/170a60a7995f73c3fc38e642.jpeg!custom.jpg"" />', 'created_at': 1583333670000, 'retweet_count': 0, 'reply_count': 0, 'fav_count': 0, 'truncated': False, 'commentId': 0, 'retweet_status_id': 0, 'symbol_id': None, 'description': ""...足。明明一个非常好的“人像留色”功能和卖点，被拍出了使用说明书教程范儿。土！！！ 如果是华为苹果三星罗永浩来拍，效果不要太骚 $小米集团-W( <span class='highlight'>01810</span> )$ 网页链接 "", 'type': '0', 'source_link': 'http://xueqiu.com/about/mobile-xueqiu', 'edited_at': None, 'pic': 'https://xqimg.imedao.com/170a60a7995f73c3fc38e642.jpeg!thumb.jpg', 'user': {'subscribeable': False, 'remark': None, 'status': 2, 'common_count': 0, 'domain': None, 'description': '如果觉得我说的不对，请用数据和事实来打我脸，而不是想当然和情绪。', 'type': '1', 'location': None, 'id': 4706271602, 'stats': None, 'url': None, 'screen_name': '小米情报官', 'following': False, 'profile': '/4706271602', 'verified': False, 'blocking': False, 'donate_count': 0, 'friends_count': 12, 'followers_count': 118, 'st_color': '1', 'follow_me': False, 'verified_description': None, 'province': '北京', 'city': '不限', 'gender': 'm', 'step': 'null', 'blog_description': None, 'verified_type': 0, 'last_status_id': 142932571, 'status_count': 171, 'recommend': None, 'intro': None, 'stock_status_count': None, 'allow_all_stock': False, 'stocks_count': None, 'name': None, 'group_ids': None, 'name_pinyin': None, 'screenname_pinyin': None, 'block_status': 0, '_source': None, 'verified_realname': False, 'verified_infos': None, 'profile_image_url': 'community/20201/1582703078929-1582703083268.jpg,community/20201/1582703078929-1582703083268.jpg!180x180.png,community/20201/1582703078929-1582703083268.jpg!50x50.png,community/20201/1582703078929-1582703083268.jpg!30x30.png', 'photo_domain': '//xavatar.imedao.com/'}, 'retweeted_status': None, 'target': '/4706271602/142932571', 'fragment': None, 'blocked': False, 'blocking': False, 'topic_pic': None, 'topic_symbol': None, 'topic_title': None, 'topic_desc': None, 'donate_count': 0, 'donate_snowcoin': 0, 'view_count': 0, 'offer': None, 'mark': 0, 'comment': None, 'expend': True, 'rqtype': None, 'rqid': 0, 'firstImg': 'https://xqimg.imedao.com/170a60a7995f73c3fc38e642.jpeg!thumb.jpg', 'card': None, 'favorited': False, 'favorited_created_at': None, 'topic_pic_thumbnail': None, 'topic_pic_headOrPad': None, 'topic_pic_thumbnail_small': None, 'canEdit': True, 'shortStatusMultiPic': False, 'timeBefore': '13分钟前', 'mark_desc': None, 'reward_count': 0, 'reward_user_count': 0, 'paid_mention': None, 'is_answer': False, 'is_refused': False, 'is_reward': True, '_source': 'sc_1:4:01810', 'pic_sizes': [{'width': 1920, 'height': 1080, 'ratio': '1.78', 'score': 1, 'setScore': True, 'setWidth': True, 'setHeight': True, 'setRatio': True}], 'like_count': 4, 'liked': False, 'promotion_pic': None, 'promotion_url': None, 'promotion_id': 0, 'quote_cards': None, 'is_ss_multi_pic': False, 'excellent_comments': None, 'is_bonus': False, 'source': 'Android'}"
142930689,6625825520,1583332679000,小米 ，余承东说你今天就消失了,"{'id': 142930689, 'user_id': 6625825520, 'title': '', 'text': '小米<a href=""http://xueqiu.com/S/01810"" target=""_blank"">$小米集团-W(<span class=\'highlight\'>01810</span>)$</a> ，余承东说你今天就消失了<br/><img class=""ke_img"" src=""https://xqimg.imedao.com/170a5fb5b04f7a53fddd36b9.jpeg!custom.jpg"" />', 'created_at': 1583332679000, 'retweet_count': 1, 'reply_count': 3, 'fav_count': 0, 'truncated': False, 'commentId': 0, 'retweet_status_id': 0, 'symbol_id': None, 'description': ""小米 $小米集团-W( <span class='highlight'>01810</span> )$ ，余承东说你今天就消失了 "", 'type': '0', 'source_link': 'http://xueqiu.com/about/mobile-xueqiu', 'edited_at': None, 'pic': 'https://xqimg.imedao.com/170a5fb5b04f7a53fddd36b9.jpeg!thumb.jpg', 'user': {'subscribeable': False, 'remark': None, 'status': 2, 'common_count': 0, 'domain': None, 'description': 'TMT分析师，关心商业、科技与创投', 'type': '1', 'location': None, 'id': 6625825520, 'stats': None, 'url': None, 'screen_name': '七寒谷', 'following': False, 'profile': '/6625825520', 'verified': False, 'blocking': False, 'donate_count': 0, 'friends_count': 68, 'followers_count': 33, 'st_color': '1', 'follow_me': False, 'verified_description': None, 'province': '不限', 'city': '未知', 'gender': 'n', 'step': 'null', 'blog_description': None, 'verified_type': 0, 'last_status_id': 142930689, 'status_count': 482, 'recommend': None, 'intro': None, 'stock_status_count': None, 'allow_all_stock': False, 'stocks_count': None, 'name': None, 'group_ids': None, 'name_pinyin': None, 'screenname_pinyin': None, 'block_status': 0, '_source': None, 'verified_realname': False, 'verified_infos': None, 'profile_image_url': 'community/20182/1519833979062-1519833979240.jpg,community/20182/1519833979062-1519833979240.jpg!180x180.png,community/20182/1519833979062-1519833979240.jpg!50x50.png,community/20182/1519833979062-1519833979240.jpg!30x30.png', 'photo_domain': '//xavatar.imedao.com/'}, 'retweeted_status': None, 'target': '/6625825520/142930689', 'fragment': None, 'blocked': False, 'blocking': False, 'topic_pic': None, 'topic_symbol': None, 'topic_title': None, 'topic_desc': None, 'donate_count': 0, 'donate_snowcoin': 0, 'view_count': 0, 'offer': None, 'mark': 0, 'comment': None, 'expend': True, 'rqtype': None, 'rqid': 0, 'firstImg': 'https://xqimg.imedao.com/170a5fb5b04f7a53fddd36b9.jpeg!thumb.jpg', 'card': None, 'favorited': False, 'favorited_created_at': None, 'topic_pic_thumbnail': None, 'topic_pic_headOrPad': None, 'topic_pic_thumbnail_small': None, 'canEdit': True, 'shortStatusMultiPic': False, 'timeBefore': '30分钟前', 'mark_desc': None, 'reward_count': 0, 'reward_user_count': 0, 'paid_mention': None, 'is_answer': False, 'is_refused': False, 'is_reward': True, '_source': 'sc_1:5:01810', 'pic_sizes': [{'width': 1080, 'height': 2340, 'ratio': '0.46', 'score': 1, 'setScore': True, 'setWidth': True, 'setHeight': True, 'setRatio': True}], 'like_count': 1, 'liked': False, 'promotion_pic': None, 'promotion_url': None, 'promotion_id': 0, 'quote_cards': None, 'is_ss_multi_pic': False, 'excellent_comments': None, 'is_bonus': False, 'source': 'Android'}"
142929670,2620249675,1583332305000,3月4日收盘后，山东道恩高分子材料股份有限公司（股票简称：道恩股份，股票代码：002838）对外发布公告称，公司股票于2020年3月2日、3月3日、3月4日连续三个交易日收盘价格涨幅偏离值累计超过20%，根据规定属于股票交易异常波动的情况。值得关注的是，2020年刚刚过去两个月，但道恩股份已经是第五次披露关于公司股票交易异常波动的公告了，不过道恩股份真正引起市场关注的还是从国内爆发疫情说起。疫情爆发后国内口罩紧缺，而作为熔喷无纺布上游产品聚丙烯熔喷料的国内最大生产企业，以及口罩布聚丙烯熔喷专用料主要生产企业，道恩股份也正式站上了风口。2月3日开盘，道恩股份的股价连续出现7个交易日涨停，成为市场最火爆的公司之一。截至3月4日收盘，道恩股份从春节开盘后起共计出现了15个涨停板，公司的股价也从2月3日11.46元的开盘，到今日收盘时已经上涨至44.33元每股，公司的总市值达到了180亿元。值得一提的是春节前，道恩股份发布了2019年度业绩预告，公司预计实现归属于上市公司股东的净利润为盈利15915.19万元–18363.68万元，同比增长30%-50%。道恩股份表示，公司业绩增加的主要原因是 2018 年 11 月完成了重大资产重组，合并报表中包含了子公司青岛海尔新材料研发有限公司的经营业绩；同时公司内部加强成本管控，增加销售收入，提高生产效率，母公司经营业绩也实现了增长。览富财经资料显示，道恩股份凭借自主研发的新型复合材料开发和应用技术，针对国内下游客户的应用需求，进行有针对性的基础研发和应用配方研发，在保证产品品质的同时，降低下游客户的采购成本、提高其采购效率，甚至根据客户需求特点进行材料配方的主动改进，实现主流客户需 求的优化满足，从而实现对行业应用的引导。自2006年以来，道恩股份先后成为一汽集团、上海大众、长城汽车、吉利汽车、海尔集团、海信集团、九阳股份等多家国内企业及其零部件配套厂商的供应商，并形成了稳定的合作关系。近年来，随着公司品牌知名度不断提升，下游诸多跨国企业与道恩股份建立起了合作关系。目前，道恩股份已经与鬼怒川（日本）、申雅密封件、库博公司（美国）、和仁宝利得（韩国）、井上华翔（日本）等全球汽车密封系统、汽车防尘罩系统行业龙头企业建立了合作关系，产品用于日产汽车、本田汽车、通用汽车、神龙汽车、现代汽车、长城汽车等汽车制造商。,"{'id': 142929670, 'user_id': 2620249675, 'title': '道恩股份：迎来春节后的第15个涨停板', 'text': '<p><a href=""http://xueqiu.com/S/SZ002838"" target=""_blank"">$道恩股份(SZ002838)$</a><a href=""http://xueqiu.com/S/01810"" target=""_blank"">$小米集团-W(<span class=\'highlight\'>01810</span>)$</a><a href=""http://xueqiu.com/S/00700"" target=""_blank"">$腾讯控股(00700)$</a></p><img class=""ke_img"" src=""https://xqimg.imedao.com/170a5f54bb8f78f3fe1b3aeb.jpeg!custom.jpg"" /><br/><p>3月4日收盘后，山东道恩高分子材料股份有限公司（股票简称：道恩股份，股票代码：002838）对外发布公告称，公司股票于2020年3月2日、3月3日、3月4日连续三个交易日收盘价格涨幅偏离值累计超过20%，根据规定属于股票交易异常波动的情况。<br/><br/>值得关注的是，2020年刚刚过去两个月，但道恩股份已经是第五次披露关于公司股票交易异常波动的公告了，不过道恩股份真正引起市场关注的还是从国内爆发疫情说起。<br/><br/>疫情爆发后国内口罩紧缺，而作为熔喷无纺布上游产品聚丙烯熔喷料的国内最大生产企业，以及口罩布聚丙烯熔喷专用料主要生产企业，道恩股份也正式站上了风口。2月3日开盘，道恩股份的股价连续出现7个交易日涨停，成为市场最火爆的公司之一。<br/><br/>截至3月4日收盘，道恩股份从春节开盘后起共计出现了15个涨停板，公司的股价也从2月3日11.46元的开盘，到今日收盘时已经上涨至44.33元每股，公司的总市值达到了180亿元。<br/><br/><br/><br/>值得一提的是春节前，道恩股份发布了2019年度业绩预告，公司预计实现归属于上市公司股东的净利润为盈利15915.19万元–18363.68万元，同比增长30%-50%。<br/><br/>道恩股份表示，公司业绩增加的主要原因是 2018 年 11 月完成了重大资产重组，合并报表中包含了子公司青岛海尔新材料研发有限公司的经营业绩；同时公司内部加强成本管控，增加销售收入，提高生产效率，母公司经营业绩也实现了增长。<br/><br/><br/><br/>览富财经资料显示，道恩股份凭借自主研发的新型复合材料开发和应用技术，针对国内下游客户的应用需求，进行有针对性的基础研发和应用配方研发，在保证产品品质的同时，降低下游客户的采购成本、提高其采购效率，甚至根据客户需求特点进行材料配方的主动改进，实现主流客户需 求的优化满足，从而实现对行业应用的引导。<br/><br/>自2006年以来，道恩股份先后成为一汽集团、上海大众、长城汽车、吉利汽车、海尔集团、海信集团、九阳股份等多家国内企业及其零部件配套厂商的供应商，并形成了稳定的合作关系。<br/><br/>近年来，随着公司品牌知名度不断提升，下游诸多跨国企业与道恩股份建立起了合作关系。目前，道恩股份已经与鬼怒川（日本）、申雅密封件、库博公司（美国）、和仁宝利得（韩国）、井上华翔（日本）等全球汽车密封系统、汽车防尘罩系统行业龙头企业建立了合作关系，产品用于日产汽车、本田汽车、通用汽车、神龙汽车、现代汽车、长城汽车等汽车制造商。</p>', 'created_at': 1583332305000, 'retweet_count': 0, 'reply_count': 0, 'fav_count': 0, 'truncated': False, 'commentId': 0, 'retweet_status_id': 0, 'symbol_id': None, 'description': ' <a href=""http://xueqiu.com/S/SZ002838"" target=""_blank"">$道恩股份(SZ002838)$</a> $小米集团-W( <span class=\'highlight\'>01810</span> )$ <a href=""http://xueqiu.com/S/00700"" target=""_blank"">$腾讯控股(00700)$</a> 3月4日收盘后，山东道恩高分子材料股份有限公司（股票简称：道恩股份，股票代码：002838）对外发布公告称，公司股票于2020年3月2日、3月3日、3月4日连续三个交易日收盘价格涨幅偏离值累计超过20%，根据规定属于股票交易异常波动的情况。 值得...', 'type': '3', 'source_link': 'http://xueqiu.com/about/mobile-xueqiu', 'edited_at': None, 'pic': 'https://xqimg.imedao.com/170a5f54bb8f78f3fe1b3aeb.jpeg!thumb.jpg', 'user': {'subscribeable': False, 'remark': None, 'status': 2, 'common_count': 0, 'domain': None, 'description': '科技信徒，误入投资领域八年，相信投资改变命运', 'type': '1', 'location': None, 'id': 2620249675, 'stats': None, 'url': None, 'screen_name': '九江水', 'following': False, 'profile': '/2620249675', 'verified': False, 'blocking': False, 'donate_count': 0, 'friends_count': 298, 'followers_count': 458, 'st_color': '1', 'follow_me': False, 'verified_description': None, 'province': '北京', 'city': '东城区', 'gender': 'n', 'step': 'null', 'blog_description': None, 'verified_type': 0, 'last_status_id': 142930946, 'status_count': 284, 'recommend': None, 'intro': None, 'stock_status_count': None, 'allow_all_stock': False, 'stocks_count': None, 'name': None, 'group_ids': None, 'name_pinyin': None, 'screenname_pinyin': None, 'block_status': 0, '_source': None, 'verified_realname': False, 'verified_infos': None, 'profile_image_url': 'community/20202/1583031899854-1583031900238.jpeg,community/20202/1583031899854-1583031900238.jpeg!180x180.png,community/20202/1583031899854-1583031900238.jpeg!50x50.png,community/20202/1583031899854-1583031900238.jpeg!30x30.png', 'photo_domain': '//xavatar.imedao.com/'}, 'retweeted_status': None, 'target': '/2620249675/142929670', 'fragment': None, 'blocked': False, 'blocking': False, 'topic_pic': None, 'topic_symbol': None, 'topic_title': None, 'topic_desc': None, 'donate_count': 0, 'donate_snowcoin': 0, 'view_count': 0, 'offer': None, 'mark': 0, 'comment': None, 'expend': True, 'rqtype': None, 'rqid': 0, 'firstImg': 'https://xqimg.imedao.com/170a5f54bb8f78f3fe1b3aeb.jpeg!thumb.jpg', 'card': None, 'favorited': False, 'favorited_created_at': None, 'topic_pic_thumbnail': None, 'topic_pic_headOrPad': None, 'topic_pic_thumbnail_small': None, 'canEdit': True, 'shortStatusMultiPic': False, 'timeBefore': '36分钟前', 'mark_desc': None, 'reward_count': 0, 'reward_user_count': 0, 'paid_mention': None, 'is_answer': False, 'is_refused': False, 'is_reward': True, '_source': 'sc_1:6:01810', 'pic_sizes': [{'width': 554, 'height': 335, 'ratio': '1.65', 'score': 1, 'setScore': True, 'setWidth': True, 'setHeight': True, 'setRatio': True}], 'like_count': 1, 'liked': False, 'promotion_pic': None, 'promotion_url': None, 'promotion_id': 0, 'quote_cards': None, 'is_ss_multi_pic': False, 'excellent_comments': None, 'is_bonus': False, 'source': 'Android'}"
142928462,2620249675,1583331864000,"国家集成电路大基金再出手2.4亿！近日，国家集成电路产业投资基金股份有限公司新增对外投资，出资2.4亿元人民币，参与成立广州兴科半导体有限公司，持股24%。天眼查数据显示，广州兴科半导体有限公司成立于2020年2月24日，是深圳市兴森快捷电路科技股份有限公司公司与科学城（广州）投资集团有限公司、国家大基金、广州兴森众城企业管理合伙企业（有限合伙）共同投资设立，注册资本为10亿元。经营范围包括集成电路封装产品设计、集成电路封装产品制造、电子元件及组件制造等。股东信息显示，公司第一大股东为A股上市公司兴森科技（002436.SZ）的运营主体兴森快捷电路，直接认缴额为4.1亿元人民币，持股比例为41%。同时，包括董事长邱醒亚在内的五人组成的广州兴森众城企业管理合伙企业（有限合伙）认缴额为1亿元人民币，持股占比10%。此外，广州兴科半导体有限公司的第二、第三大股东分别为科学城（广州）投资集团有限公司、国家集成电路产业投资基金股份有限公司，二者认缴额为2.5亿元、2.4亿元，持股占比分别为25%、24%。兴森科技半导体封装基地正式动工据悉，兴森科技半导体封装基板产业基地项目投资16亿元人民币，规划产能为30000平米/月的集成电路封装基板和15000平米/月的类载板，是国家集成电路产业基金在广州的第一个项目。2月28日，兴森科技半导体封装项目正式举行破土动工仪式。这一项目将致力于解决我国半导体产业链卡脖子问题，提升自主创新能力，有效弥补我国半导体产业链的短板。图片源自兴森科技其实，半导体封装基板也称封装载板，是集成电路封装的最重要原物料之一。主要功能是作为载体承载集成电路芯片，并以封装基板内部线路连接晶圆与印刷电路板（PCB）之间的讯号。能够保护电路，固定线路并导散余热，是封装过程中最重要的原材料。16个交易日股价暴涨90%2019年经济润涨逾4成3月3日，兴森科技在互动平台表示，公司不生产口罩。生产的PCB产品广泛应用于红外检测、超声仪器、安防AI等医疗设备，公司已通过ISO13485医疗器械质量管理体系认证，与国内外医疗设备、医疗器械龙头企业均有业务合作关系。目前正全力赶制电子测温仪等订单。受上述消息影响，今日兴森科技低开高走，一度直线拉升至7.75%，随后区间震荡，截至收盘，该股涨5.97%，报14.90元，成交额25.9亿元，最新总市值达222亿元。此前兴森科技曾实现11连涨，并连续创出阶段高点。2月4日至2月25日，16个交易日股价大涨90.66%，市值暴涨126亿元。据悉，兴森科技是是国内最大的印制电路样板小批量板快件制造商，一直致力于为国内外高科技电子企业和科研单位服务,产品广泛运用于通信、网络、工业控制、计算机应用、国防军工、航天、医疗等行业领域。公司先后成为华为、中兴核心快件样板供应商，并与中车、浪潮、西门子、三星、高通等超过五千多家海内外知名品牌公司及电子研发类企业建立了良好的合作关系。公司的主营业务是印制电路板制造服务，正在打造板卡业务、半导体业务、一站式业务。2月25日，兴森科技发布2019年度业绩快报，其中披露，2019年度兴森科技营业总收入同比增长9.19%，达到37.92亿元。该公司称，销售收入保持平稳增长，主要来自半导体测试板业务、IC封装基板业务以及SMT业务收入增长。另外，归属于上市公司股东的净利润为3.03亿元，同比增长41.09%。该公司对原因分析包括，美国Harbor、宜兴硅谷、英国Exception等子公司经营情况进一步改善；子公司上海泽丰收入取得较大幅度的增长，盈利增加；公司实施的降本增效措施效果呈现，成本费用率有所下降，盈利能力提升。在2019半年报可以看到，兴森科技的主营业务分行业而言主要为PCB业务与半导体业务，2019上半年PCB业务占营业总收入的77%，整体增长平稳，子公司实现盈利目标。同时，公司半导体业务占总营收的27%左右，业绩方面取得较大进展，毛利率明显提高，盈利能力显著改善。市场方面，公司在前期在5G产品上面布局比较全面，今年国内5G牌照的发放促进了相关需求的快速增长。国家战略层面，国内半导体产业的迅速发展也催生了配套产业链的快速本土化。机构：看好半导体未来趋势其实，兴森科技年后涨势亮眼，一定程度上是搭上了半导体的“顺风车”。鼠年开市以来，半导体板块可谓是气势如虹，势不可挡。但随后有所调整，截至3月4日收盘，半导体指数回落至3729点，单日跌幅达5.9%，经历大涨后跌连续下跌，市场担忧情绪升级。受新冠疫情全球迅速蔓延的影响，全球半导体的供应及上下游将受影响。申港证券研报表示，在全球半导体材料领域，日本占据绝对主导地位，在半导体制造过程包含的19种核心材料中，日本市占率超过50%份额的材料就占到了14种，在全球半导体材料领域处于绝对领先地位。在硅片领域，日本的两家企业就占据全球53%的市场份额，另外，日本的JSR、东京应化、信越化学及富士电子四家企业占据了全球70%以上的光刻胶市场份额，处于市场垄断地位。光刻胶属于高技术壁垒材料，生产工艺复杂，纯度要求高，需要长期的技术积累。如果疫情导致了原材料供应受限，将对全球半导体领域造成重大影响。平安证券表示，日韩疫情发酵，带来全球半导体产业链的不确定性。分析表示，疫情在日韩流行超出预期，考虑到日韩在整个电子产业供应中的关键位置：韩国DRAM市占率超过70%，日本在CIS、硅晶圆材料、光罩、靶材等领域市占率超过50%，电子行业的供应风险上升；且前期科技板块持续上涨，导致申万电子行业指数本周下跌幅度较大。认为，虽然短期来看疫情负面影响不可避免，但随着疫情解除，有望迎来报复式反弹。在不忽略短期风险的前提下，仍是看好5G趋势下半导体景气向上的大趋势。","{'id': 142928462, 'user_id': 2620249675, 'title': '股价16天暴涨90％！这家公司凭啥这么牛？', 'text': '<p><a href=""http://xueqiu.com/S/SZ002436"" target=""_blank"">$兴森科技(SZ002436)$</a><a href=""http://xueqiu.com/S/01810"" target=""_blank"">$小米集团-W(<span class=\'highlight\'>01810</span>)$</a><a href=""http://xueqiu.com/S/00700"" target=""_blank"">$腾讯控股(00700)$</a></p><img class=""ke_img"" src=""https://xqimg.imedao.com/170a5ee4c1ff5d63fc4ff5b6.jpeg!custom.jpg"" /><br/><p>国家集成电路大基金再出手2.4亿！<br/><br/>近日，国家集成电路产业投资基金股份有限公司新增对外投资，出资2.4亿元人民币，参与成立广州兴科半导体有限公司，持股24%。<br/><br/>天眼查数据显示，广州兴科半导体有限公司成立于2020年2月24日，是深圳市兴森快捷电路科技股份有限公司公司与科学城（广州）投资集团有限公司、国家大基金、广州兴森众城企业管理合伙企业（有限合伙）共同投资设立，注册资本为10亿元。经营范围包括集成电路封装产品设计、集成电路封装产品制造、电子元件及组件制造等。<br/><br/><br/><br/>股东信息显示，公司第一大股东为A股上市公司兴森科技（002436.SZ）的运营主体兴森快捷电路，直接认缴额为4.1亿元人民币，持股比例为41%。同时，包括董事长邱醒亚在内的五人组成的广州兴森众城企业管理合伙企业（有限合伙）认缴额为1亿元人民币，持股占比10%。<br/><br/><br/><br/>此外，广州兴科半导体有限公司的第二、第三大股东分别为科学城（广州）投资集团有限公司、国家集成电路产业投资基金股份有限公司，二者认缴额为2.5亿元、2.4亿元，持股占比分别为25%、24%。<br/><br/><br/><br/>兴森科技半导体封装基地正式动工<br/><br/>据悉，兴森科技半导体封装基板产业基地项目投资16亿元人民币，规划产能为30000平米/月的集成电路封装基板和15000平米/月的类载板，是国家集成电路产业基金在广州的第一个项目。<br/><br/>2月28日，兴森科技半导体封装项目正式举行破土动工仪式。这一项目将致力于解决我国半导体产业链卡脖子问题，提升自主创新能力，有效弥补我国半导体产业链的短板。<br/><br/><br/><br/>图片源自兴森科技<br/><br/>其实，半导体封装基板也称封装载板，是集成电路封装的最重要原物料之一。主要功能是作为载体承载集成电路芯片，并以封装基板内部线路连接晶圆与印刷电路板（PCB）之间的讯号。能够保护电路，固定线路并导散余热，是封装过程中最重要的原材料。<br/><br/>16个交易日股价暴涨90%<br/><br/>2019年经济润涨逾4成<br/><br/>3月3日，兴森科技在互动平台表示，公司不生产口罩。生产的PCB产品广泛应用于红外检测、超声仪器、安防AI等医疗设备，公司已通过ISO13485医疗器械质量管理体系认证，与国内外医疗设备、医疗器械龙头企业均有业务合作关系。目前正全力赶制电子测温仪等订单。<br/><br/>受上述消息影响，今日兴森科技低开高走，一度直线拉升至7.75%，随后区间震荡，截至收盘，该股涨5.97%，报14.90元，成交额25.9亿元，最新总市值达222亿元。<br/><br/>此前兴森科技曾实现11连涨，并连续创出阶段高点。2月4日至2月25日，16个交易日股价大涨90.66%，市值暴涨126亿元。<br/><br/><br/><br/>据悉，兴森科技是是国内最大的印制电路样板小批量板快件制造商，一直致力于为国内外高科技电子企业和科研单位服务,产品广泛运用于通信、网络、工业控制、计算机应用、国防军工、航天、医疗等行业领域。公司先后成为华为、中兴核心快件样板供应商，并与中车、浪潮、西门子、三星、高通等超过五千多家海内外知名品牌公司及电子研发类企业建立了良好的合作关系。公司的主营业务是印制电路板制造服务，正在打造板卡业务、半导体业务、一站式业务。<br/><br/>2月25日，兴森科技发布2019年度业绩快报，其中披露，2019年度兴森科技营业总收入同比增长9.19%，达到37.92亿元。该公司称，销售收入保持平稳增长，主要来自半导体测试板业务、IC封装基板业务以及SMT业务收入增长。<br/><br/><br/><br/>另外，归属于上市公司股东的净利润为3.03亿元，同比增长41.09%。该公司对原因分析包括，美国Harbor、宜兴硅谷、英国Exception等子公司经营情况进一步改善；子公司上海泽丰收入取得较大幅度的增长，盈利增加；公司实施的降本增效措施效果呈现，成本费用率有所下降，盈利能力提升。<br/><br/>在2019半年报可以看到，兴森科技的主营业务分行业而言主要为PCB业务与半导体业务，2019上半年PCB业务占营业总收入的77%，整体增长平稳，子公司实现盈利目标。同时，公司半导体业务占总营收的27%左右，业绩方面取得较大进展，毛利率明显提高，盈利能力显著改善。市场方面，公司在前期在5G产品上面布局比较全面，今年国内5G牌照的发放促进了相关需求的快速增长。国家战略层面，国内半导体产业的迅速发展也催生了配套产业链的快速本土化。<br/><br/>机构：看好半导体未来趋势<br/><br/>其实，兴森科技年后涨势亮眼，一定程度上是搭上了半导体的“顺风车”。<br/><br/>鼠年开市以来，半导体板块可谓是气势如虹，势不可挡。但随后有所调整，截至3月4日收盘，半导体指数回落至3729点，单日跌幅达5.9%，经历大涨后跌连续下跌，市场担忧情绪升级。<br/><br/>受新冠疫情全球迅速蔓延的影响，全球半导体的供应及上下游将受影响。申港证券研报表示，在全球半导体材料领域，日本占据绝对主导地位，在半导体制造过程包含的19种核心材料中，日本市占率超过50%份额的材料就占到了14种，在全球半导体材料领域处于绝对领先地位。在硅片领域，日本的两家企业就占据全球53%的市场份额，另外，日本的JSR、东京应化、信越化学及富士电子四家企业占据了全球70%以上的光刻胶市场份额，处于市场垄断地位。光刻胶属于高技术壁垒材料，生产工艺复杂，纯度要求高，需要长期的技术积累。如果疫情导致了原材料供应受限，将对全球半导体领域造成重大影响。<br/><br/>平安证券表示，日韩疫情发酵，带来全球半导体产业链的不确定性。分析表示，疫情在日韩流行超出预期，考虑到日韩在整个电子产业供应中的关键位置：韩国DRAM市占率超过70%，日本在CIS、硅晶圆材料、光罩、靶材等领域市占率超过50%，电子行业的供应风险上升；且前期科技板块持续上涨，导致申万电子行业指数本周下跌幅度较大。认为，虽然短期来看疫情负面影响不可避免，但随着疫情解除，有望迎来报复式反弹。在不忽略短期风险的前提下，仍是看好5G趋势下半导体景气向上的大趋势。</p>', 'created_at': 1583331864000, 'retweet_count': 0, 'reply_count': 0, 'fav_count': 0, 'truncated': False, 'commentId': 0, 'retweet_status_id': 0, 'symbol_id': None, 'description': ' <a href=""http://xueqiu.com/S/SZ002436"" target=""_blank"">$兴森科技(SZ002436)$</a> $小米集团-W( <span class=\'highlight\'>01810</span> )$ <a href=""http://xueqiu.com/S/00700"" target=""_blank"">$腾讯控股(00700)$</a> 国家集成电路大基金再出手2.4亿！ 近日，国家集成电路产业投资基金股份有限公司新增对外投资，出资2.4亿元人民币，参与成立广州兴科半导体有限公司，持股24%。 天眼查数据显示，广州兴科半导体有限公司成立于2020年2月24日，是深圳市兴森快捷...', 'type': '3', 'source_link': 'http://xueqiu.com/about/mobile-xueqiu', 'edited_at': None, 'pic': 'https://xqimg.imedao.com/170a5ee4c1ff5d63fc4ff5b6.jpeg!thumb.jpg', 'user': {'subscribeable': False, 'remark': None, 'status': 2, 'common_count': 0, 'domain': None, 'description': '科技信徒，误入投资领域八年，相信投资改变命运', 'type': '1', 'location': None, 'id': 2620249675, 'stats': None, 'url': None, 'screen_name': '九江水', 'following': False, 'profile': '/2620249675', 'verified': False, 'blocking': False, 'donate_count': 0, 'friends_count': 298, 'followers_count': 458, 'st_color': '1', 'follow_me': False, 'verified_description': None, 'province': '北京', 'city': '东城区', 'gender': 'n', 'step': 'null', 'blog_description': None, 'verified_type': 0, 'last_status_id': 142930946, 'status_count': 284, 'recommend': None, 'intro': None, 'stock_status_count': None, 'allow_all_stock': False, 'stocks_count': None, 'name': None, 'group_ids': None, 'name_pinyin': None, 'screenname_pinyin': None, 'block_status': 0, '_source': None, 'verified_realname': False, 'verified_infos': None, 'profile_image_url': 'community/20202/1583031899854-1583031900238.jpeg,community/20202/1583031899854-1583031900238.jpeg!180x180.png,community/20202/1583031899854-1583031900238.jpeg!50x50.png,community/20202/1583031899854-1583031900238.jpeg!30x30.png', 'photo_domain': '//xavatar.imedao.com/'}, 'retweeted_status': None, 'target': '/2620249675/142928462', 'fragment': None, 'blocked': False, 'blocking': False, 'topic_pic': None, 'topic_symbol': None, 'topic_title': None, 'topic_desc': None, 'donate_count': 0, 'donate_snowcoin': 0, 'view_count': 0, 'offer': None, 'mark': 0, 'comment': None, 'expend': True, 'rqtype': None, 'rqid': 0, 'firstImg': 'https://xqimg.imedao.com/170a5ee4c1ff5d63fc4ff5b6.jpeg!thumb.jpg', 'card': None, 'favorited': False, 'favorited_created_at': None, 'topic_pic_thumbnail': None, 'topic_pic_headOrPad': None, 'topic_pic_thumbnail_small': None, 'canEdit': True, 'shortStatusMultiPic': False, 'timeBefore': '44分钟前', 'mark_desc': None, 'reward_count': 0, 'reward_user_count': 0, 'paid_mention': None, 'is_answer': False, 'is_refused': False, 'is_reward': True, '_source': 'sc_1:7:01810', 'pic_sizes': [{'width': 554, 'height': 354, 'ratio': '1.56', 'score': 7, 'setScore': True, 'setWidth': True, 'setHeight': True, 'setRatio': True}], 'like_count': 1, 'liked': False, 'promotion_pic': None, 'promotion_url': None, 'promotion_id': 0, 'quote_cards': None, 'is_ss_multi_pic': False, 'excellent_comments': None, 'is_bonus': False, 'source': 'Android'}"
142927482,2620249675,1583331164000,3月4日消息，黑鲨3在京东开启预约。不到一天时间，黑鲨3在京东预约量突破360万，人气火爆。值得注意的是，玩家除了参与预约之外，也可以支付定金直接下单，3月6日上午10点正式首销，售价3499元（8GB+128GB）。作为迄今为止性价比最高的骁龙865机型，黑鲨3最大的看点之一是散热。据悉，黑鲨3采用“三明治”液冷，所有热源全覆盖。其液冷管总长度达到了约186mm，相较上一代，热源散热面积提升100%，散热能力提升50%。规格方面，它采用6.67英寸三星AMOLED显示屏，刷新率为90Hz，触控采样率达到了270Hz，搭载高通骁龙865旗舰平台，最高配备LPDDR5内存及UFS 3.0闪存，电池容量为4720mAh，支持65W极速闪充，38分钟充电100%。不仅如此，它搭载MEMC运动补偿技术，将低帧率内容提升至最高90帧每秒，呈现真90Hz流畅体验，突破游戏帧率限制，实现游戏内更高帧率显示。此外，黑鲨3搭载黑鲨GameEngine和腾讯SolarCore双引擎，它们能对处理器、网络、内存等系统资源进行协同调度，给玩家提供更好的游戏性能体验，游戏启动和加载速度提升10%，游戏帧率波动降低50%。,"{'id': 142927482, 'user_id': 2620249675, 'title': '不到一天预约量突破360万 黑鲨3火了：3499元', 'text': '<p><a href=""http://xueqiu.com/S/01810"" target=""_blank"">$小米集团-W(<span class=\'highlight\'>01810</span>)$</a><a href=""http://xueqiu.com/S/00700"" target=""_blank"">$腾讯控股(00700)$</a><a href=""http://xueqiu.com/S/JD"" target=""_blank"">$京东(JD)$</a></p><img class=""ke_img"" src=""https://xqimg.imedao.com/170a5e436a0f7733fec2f185.jpeg!custom.jpg"" /><br/><p><br/>3月4日消息，黑鲨3在京东开启预约。<br/><br/>不到一天时间，黑鲨3在京东预约量突破360万，人气火爆。<br/><br/>值得注意的是，玩家除了参与预约之外，也可以支付定金直接下单，3月6日上午10点正式首销，售价3499元（8GB+128GB）。<br/><br/><br/><br/>作为迄今为止性价比最高的骁龙865机型，黑鲨3最大的看点之一是散热。<br/><br/>据悉，黑鲨3采用“三明治”液冷，所有热源全覆盖。其液冷管总长度达到了约186mm，相较上一代，热源散热面积提升100%，散热能力提升50%。<br/><br/>规格方面，它采用6.67英寸三星AMOLED显示屏，刷新率为90Hz，触控采样率达到了270Hz，搭载高通骁龙865旗舰平台，最高配备LPDDR5内存及UFS 3.0闪存，电池容量为4720mAh，支持65W极速闪充，38分钟充电100%。<br/><br/>不仅如此，它搭载MEMC运动补偿技术，将低帧率内容提升至最高90帧每秒，呈现真90Hz流畅体验，突破游戏帧率限制，实现游戏内更高帧率显示。<br/><br/>此外，黑鲨3搭载黑鲨GameEngine和腾讯SolarCore双引擎，它们能对处理器、网络、内存等系统资源进行协同调度，给玩家提供更好的游戏性能体验，游戏启动和加载速度提升10%，游戏帧率波动降低50%。<br/></p>', 'created_at': 1583331164000, 'retweet_count': 0, 'reply_count': 2, 'fav_count': 0, 'truncated': False, 'commentId': 0, 'retweet_status_id': 0, 'symbol_id': None, 'description': ' $小米集团-W( <span class=\'highlight\'>01810</span> )$ <a href=""http://xueqiu.com/S/00700"" target=""_blank"">$腾讯控股(00700)$</a> <a href=""http://xueqiu.com/S/JD"" target=""_blank"">$京东(JD)$</a> 3月4日消息，黑鲨3在京东开启预约。 不到一天时间，黑鲨3在京东预约量突破360万，人气火爆。 值得注意的是，玩家除了参与预约之外，也可以支付定金直接下单，3月6日上午10点正式首销，售价3499元（8GB+128GB）。 作为迄今为止性价比最高的骁龙865机型，...', 'type': '3', 'source_link': 'http://xueqiu.com/about/mobile-xueqiu', 'edited_at': None, 'pic': 'https://xqimg.imedao.com/170a5e436a0f7733fec2f185.jpeg!thumb.jpg', 'user': {'subscribeable': False, 'remark': None, 'status': 2, 'common_count': 0, 'domain': None, 'description': '科技信徒，误入投资领域八年，相信投资改变命运', 'type': '1', 'location': None, 'id': 2620249675, 'stats': None, 'url': None, 'screen_name': '九江水', 'following': False, 'profile': '/2620249675', 'verified': False, 'blocking': False, 'donate_count': 0, 'friends_count': 298, 'followers_count': 458, 'st_color': '1', 'follow_me': False, 'verified_description': None, 'province': '北京', 'city': '东城区', 'gender': 'n', 'step': 'null', 'blog_description': None, 'verified_type': 0, 'last_status_id': 142930946, 'status_count': 284, 'recommend': None, 'intro': None, 'stock_status_count': None, 'allow_all_stock': False, 'stocks_count': None, 'name': None, 'group_ids': None, 'name_pinyin': None, 'screenname_pinyin': None, 'block_status': 0, '_source': None, 'verified_realname': False, 'verified_infos': None, 'profile_image_url': 'community/20202/1583031899854-1583031900238.jpeg,community/20202/1583031899854-1583031900238.jpeg!180x180.png,community/20202/1583031899854-1583031900238.jpeg!50x50.png,community/20202/1583031899854-1583031900238.jpeg!30x30.png', 'photo_domain': '//xavatar.imedao.com/'}, 'retweeted_status': None, 'target': '/2620249675/142927482', 'fragment': None, 'blocked': False, 'blocking': False, 'topic_pic': None, 'topic_symbol': None, 'topic_title': None, 'topic_desc': None, 'donate_count': 0, 'donate_snowcoin': 0, 'view_count': 0, 'offer': None, 'mark': 0, 'comment': None, 'expend': True, 'rqtype': None, 'rqid': 0, 'firstImg': 'https://xqimg.imedao.com/170a5e436a0f7733fec2f185.jpeg!thumb.jpg', 'card': None, 'favorited': False, 'favorited_created_at': None, 'topic_pic_thumbnail': None, 'topic_pic_headOrPad': None, 'topic_pic_thumbnail_small': None, 'canEdit': True, 'shortStatusMultiPic': False, 'timeBefore': '55分钟前', 'mark_desc': None, 'reward_count': 0, 'reward_user_count': 0, 'paid_mention': None, 'is_answer': False, 'is_refused': False, 'is_reward': True, '_source': 'sc_1:8:01810', 'pic_sizes': [{'width': 522, 'height': 697, 'ratio': '0.75', 'score': 2, 'setScore': True, 'setWidth': True, 'setHeight': True, 'setRatio': True}], 'like_count': 5, 'liked': False, 'promotion_pic': None, 'promotion_url': None, 'promotion_id': 0, 'quote_cards': None, 'is_ss_multi_pic': False, 'excellent_comments': None, 'is_bonus': False, 'source': 'Android'}"
142927192,2620249675,1583330932000,新冠病毒肆虐全球，市场随之剧烈波动，各国央行坐不住了！今天，澳大利亚联储公布利率决议，将3月现金利率下调25个基点至0.50％，创纪录新低。这还是只开始，此前关于“放水”的声明也是一个接着一个——美国：正密切监视事态发展及对经济前景的影响，承诺将采取适当行动来支持经济。日本：已向金融机构提供5000亿日元、期限两周的资金。欧洲：密切关注新冠肺炎疫情发展影响，将随时准备采取适当的针对性措施。……世界主要经济体面对疫情影响，正绞尽脑汁随时进行紧急操作。全球大“放水”模式要开启了吗？中国央行会不会跟？更关键的是，怎么影响你我的钱袋子？Q＆A会跟着“放水”吗？我国货币政策还是“以我为主”光大银行金融市场部分析师周茂华认为，海外宽松加码，一定程度将扩大我国货币政策逆周期调节空间。这不，最近，人民银行已经引导了贷款利率下调。根据此前预告，下一步，央行还将通过普惠金融定向降准考核，继续释放资金。不过，周茂华也说，我国货币政策调整仍将“以我为主”，货币政策协同财政、产业政策，加强逆周期调节，确保今年各项社会经济目标完成。交通银行金融研究中心首席研究员唐建伟认为，央行目前对于货币政策的态度依然维持稳健，并重视逆周期调节力度以及流动性去向结构的优化。中国民生银行首席研究员温彬也认为，由于通胀位于高位，可能制约央行通过下调存款基准利率，降低银行成本。Q＆A股市会涨吗？低估值周期板块受追捧一提到“放水”，最先想到的就是股市。是不是各国央行一旦开闸“放水”，股市首先要猛涨啊？确实有这个可能。受“放水”预期影响，3月2日，美国道琼斯指数暴涨5.09％，创十年多以来最大涨幅！同样，3月3日一开盘，A股三大指数也集体高开。招商基金首席宏观策略分析师姚爽认为，相关声明引发了市场对于对冲政策继续加码的预期，低估值的周期板块受到追捧。姚爽认为，目前市场估值仍处于历史中枢以下，疫情的影响仍宜定义为短期冲击，市场整体的下行风险不大。周茂华表示，短期国内股市受经济数据与海外市场波动影响，但国内疫情基本得到有效控制，监管层采取的政策组合将有效对冲疫情影响，国内股市仍处于中长期结构性慢牛轨道。Q＆A房价会升吗？对楼市影响不会太大货币政策宽松，对楼市会是利好吗？毕竟，市场上流通的钱多了，终归要有渠道去吸收，楼市自然是其中之一。还有些聪明的网友，甚至想借着疫情，在楼市抄一把底。其实，这些不切实际的想法还是打住为好。在近日存量浮动利率贷款定价基准转换启动之际，央行已经明确，存量商业性个人住房贷款在转换时点的利率水平保持不变。事实上，从央行一直以来的操作不难看出，坚守“房住不炒”的定位，不向房地产市场发出错误信号，是一以贯之的做法。周茂华认为，国内楼市稳是主基调，“房住不炒”基本定位不会动摇，管理层也多次表态，“不将房地产作为短期刺激经济的手段。”因此，全球央行“放水”，对我国楼市影响不会太大。Q＆A票子会贬值不？不仅不会贬值还可能升值如果开启新一轮“放水”？我们手里的票子会贬值不？人民银行副行长刘国强认为，汇率走势取决于经济基本面。我国经济长期向好的基本面没变，外汇储备充足，本外币利差仍处于合适区间，这些都为人民币汇率提供了根本支撑。也就是说，人民币不仅不会贬值，说不定还能变得更值钱。东方金诚国际信用评级有限公司分析师谢文杰表示，人民币并不全然会承受明显贬值压力。在基本面支撑下，预计短期内人民币汇率整体波动幅度比较有限。综合各类影响因素，东方金诚预计，短期内人民币对一篮子货币汇率指数还将出现一定幅度升值。此外，交银宏观团队认为，预计3月份人民币汇率基本稳定，或小幅升值。,"{'id': 142927192, 'user_id': 2620249675, 'title': '全球大“放水”！房价怎么走，股市能涨吗，票子会贬值不？', 'text': '<p><a href=""http://xueqiu.com/S/01810"" target=""_blank"">$小米集团-W(<span class=\'highlight\'>01810</span>)$</a><a href=""http://xueqiu.com/S/SH600036"" target=""_blank"">$招商银行(SH600036)$</a><a href=""http://xueqiu.com/S/SH600030"" target=""_blank"">$中信证券(SH600030)$</a></p><img class=""ke_img"" src=""https://xqimg.imedao.com/170a5e049ddf6ed3fd672d54.jpeg!custom.jpg"" /><br/><p>新冠病毒肆虐全球，市场随之剧烈波动，各国央行坐不住了！<br/><br/>今天，澳大利亚联储公布利率决议，将3月现金利率下调25个基点至0.50％，创纪录新低。<br/><br/>这还是只开始，此前关于“放水”的声明也是一个接着一个——<br/><br/>美国：正密切监视事态发展及对经济前景的影响，承诺将采取适当行动来支持经济。日本：已向金融机构提供5000亿日元、期限两周的资金。<br/><br/>欧洲：密切关注新冠肺炎疫情发展影响，将随时准备采取适当的针对性措施。<br/><br/>……<br/><br/>世界主要经济体面对疫情影响，正绞尽脑汁随时进行紧急操作。<br/><br/>全球大“放水”模式要开启了吗？中国央行会不会跟？更关键的是，怎么影响你我的钱袋子？<br/><br/><br/><br/>Q＆A<br/><br/>会跟着“放水”吗？<br/><br/>我国货币政策还是“以我为主”<br/><br/>光大银行金融市场部分析师周茂华认为，海外宽松加码，一定程度将扩大我国货币政策逆周期调节空间。<br/><br/>这不，最近，人民银行已经引导了贷款利率下调。<br/><br/>根据此前预告，下一步，央行还将通过普惠金融定向降准考核，继续释放资金。<br/><br/>不过，周茂华也说，我国货币政策调整仍将“以我为主”，货币政策协同财政、产业政策，加强逆周期调节，确保今年各项社会经济目标完成。<br/><br/>交通银行金融研究中心首席研究员唐建伟认为，央行目前对于货币政策的态度依然维持稳健，并重视逆周期调节力度以及流动性去向结构的优化。<br/><br/>中国民生银行首席研究员温彬也认为，由于通胀位于高位，可能制约央行通过下调存款基准利率，降低银行成本。<br/><br/><br/><br/>Q＆A<br/><br/>股市会涨吗？<br/><br/>低估值周期板块受追捧<br/><br/>一提到“放水”，最先想到的就是股市。<br/><br/>是不是各国央行一旦开闸“放水”，股市首先要猛涨啊？<br/><br/>确实有这个可能。受“放水”预期影响，3月2日，美国道琼斯指数暴涨5.09％，创十年多以来最大涨幅！<br/><br/>同样，3月3日一开盘，A股三大指数也集体高开。<br/><br/>招商基金首席宏观策略分析师姚爽认为，相关声明引发了市场对于对冲政策继续加码的预期，低估值的周期板块受到追捧。<br/><br/>姚爽认为，目前市场估值仍处于历史中枢以下，疫情的影响仍宜定义为短期冲击，市场整体的下行风险不大。<br/><br/>周茂华表示，短期国内股市受经济数据与海外市场波动影响，但国内疫情基本得到有效控制，监管层采取的政策组合将有效对冲疫情影响，国内股市仍处于中长期结构性慢牛轨道。<br/><br/><br/><br/>Q＆A<br/><br/>房价会升吗？<br/><br/>对楼市影响不会太大<br/><br/>货币政策宽松，对楼市会是利好吗？<br/><br/>毕竟，市场上流通的钱多了，终归要有渠道去吸收，楼市自然是其中之一。<br/><br/>还有些聪明的网友，甚至想借着疫情，在楼市抄一把底。<br/><br/>其实，这些不切实际的想法还是打住为好。<br/><br/>在近日存量浮动利率贷款定价基准转换启动之际，央行已经明确，存量商业性个人住房贷款在转换时点的利率水平保持不变。<br/><br/>事实上，从央行一直以来的操作不难看出，坚守“房住不炒”的定位，不向房地产市场发出错误信号，是一以贯之的做法。<br/><br/>周茂华认为，国内楼市稳是主基调，“房住不炒”基本定位不会动摇，管理层也多次表态，“不将房地产作为短期刺激经济的手段。”<br/><br/>因此，全球央行“放水”，对我国楼市影响不会太大。<br/><br/><br/><br/>Q＆A<br/><br/>票子会贬值不？<br/><br/>不仅不会贬值还可能升值<br/><br/>如果开启新一轮“放水”？我们手里的票子会贬值不？<br/><br/>人民银行副行长刘国强认为，汇率走势取决于经济基本面。我国经济长期向好的基本面没变，外汇储备充足，本外币利差仍处于合适区间，这些都为人民币汇率提供了根本支撑。<br/><br/>也就是说，人民币不仅不会贬值，说不定还能变得更值钱。<br/><br/>东方金诚国际信用评级有限公司分析师谢文杰表示，人民币并不全然会承受明显贬值压力。在基本面支撑下，预计短期内人民币汇率整体波动幅度比较有限。<br/><br/>综合各类影响因素，东方金诚预计，短期内人民币对一篮子货币汇率指数还将出现一定幅度升值。<br/><br/>此外，交银宏观团队认为，预计3月份人民币汇率基本稳定，或小幅升值。</p>', 'created_at': 1583330932000, 'retweet_count': 0, 'reply_count': 0, 'fav_count': 0, 'truncated': False, 'commentId': 0, 'retweet_status_id': 0, 'symbol_id': None, 'description': ' $小米集团-W( <span class=\'highlight\'>01810</span> )$ <a href=""http://xueqiu.com/S/SH600036"" target=""_blank"">$招商银行(SH600036)$</a> <a href=""http://xueqiu.com/S/SH600030"" target=""_blank"">$中信证券(SH600030)$</a> 新冠病毒肆虐全球，市场随之剧烈波动，各国央行坐不住了！ 今天，澳大利亚联储公布利率决议，将3月现金利率下调25个基点至0.50％，创纪录新低。 这还是只开始，此前关于“放水”的声明也是一个接着一个—— 美国：正密切监视事态发展及对...', 'type': '3', 'source_link': 'http://xueqiu.com/about/mobile-xueqiu', 'edited_at': None, 'pic': 'https://xqimg.imedao.com/170a5e049ddf6ed3fd672d54.jpeg!thumb.jpg', 'user': {'subscribeable': False, 'remark': None, 'status': 2, 'common_count': 0, 'domain': None, 'description': '科技信徒，误入投资领域八年，相信投资改变命运', 'type': '1', 'location': None, 'id': 2620249675, 'stats': None, 'url': None, 'screen_name': '九江水', 'following': False, 'profile': '/2620249675', 'verified': False, 'blocking': False, 'donate_count': 0, 'friends_count': 298, 'followers_count': 458, 'st_color': '1', 'follow_me': False, 'verified_description': None, 'province': '北京', 'city': '东城区', 'gender': 'n', 'step': 'null', 'blog_description': None, 'verified_type': 0, 'last_status_id': 142930946, 'status_count': 284, 'recommend': None, 'intro': None, 'stock_status_count': None, 'allow_all_stock': False, 'stocks_count': None, 'name': None, 'group_ids': None, 'name_pinyin': None, 'screenname_pinyin': None, 'block_status': 0, '_source': None, 'verified_realname': False, 'verified_infos': None, 'profile_image_url': 'community/20202/1583031899854-1583031900238.jpeg,community/20202/1583031899854-1583031900238.jpeg!180x180.png,community/20202/1583031899854-1583031900238.jpeg!50x50.png,community/20202/1583031899854-1583031900238.jpeg!30x30.png', 'photo_domain': '//xavatar.imedao.com/'}, 'retweeted_status': None, 'target': '/2620249675/142927192', 'fragment': None, 'blocked': False, 'blocking': False, 'topic_pic': None, 'topic_symbol': None, 'topic_title': None, 'topic_desc': None, 'donate_count': 0, 'donate_snowcoin': 0, 'view_count': 0, 'offer': None, 'mark': 0, 'comment': None, 'expend': True, 'rqtype': None, 'rqid': 0, 'firstImg': 'https://xqimg.imedao.com/170a5e049ddf6ed3fd672d54.jpeg!thumb.jpg', 'card': None, 'favorited': False, 'favorited_created_at': None, 'topic_pic_thumbnail': None, 'topic_pic_headOrPad': None, 'topic_pic_thumbnail_small': None, 'canEdit': True, 'shortStatusMultiPic': False, 'timeBefore': '59分钟前', 'mark_desc': None, 'reward_count': 0, 'reward_user_count': 0, 'paid_mention': None, 'is_answer': False, 'is_refused': False, 'is_reward': True, '_source': 'sc_1:9:01810', 'pic_sizes': [{'width': 580, 'height': 690, 'ratio': '0.84', 'score': 0, 'setScore': True, 'setWidth': True, 'setHeight': True, 'setRatio': True}], 'like_count': 2, 'liked': False, 'promotion_pic': None, 'promotion_url': None, 'promotion_id': 0, 'quote_cards': None, 'is_ss_multi_pic': False, 'excellent_comments': None, 'is_bonus': False, 'source': 'Android'}"
142926821,4706271602,1583330669000,为弥补疫情对经济影响，5G网络建设应该会加速，5G手机普及速度也会加快。对小米这样的0库存、纯5G手机制造商是利好,"{'id': 142926821, 'user_id': 4706271602, 'title': '', 'text': '为弥补疫情对经济影响，5G网络建设应该会加速，5G手机普及速度也会加快。对小米这样的0库存、纯5G手机制造商是利好<a href=""http://xueqiu.com/S/01810"" target=""_blank"">$小米集团-W(<span class=\'highlight\'>01810</span>)$</a>', 'created_at': 1583330669000, 'retweet_count': 0, 'reply_count': 0, 'fav_count': 0, 'truncated': False, 'commentId': 0, 'retweet_status_id': 0, 'symbol_id': None, 'description': ""为弥补疫情对经济影响，5G网络建设应该会加速，5G手机普及速度也会加快。对小米这样的0库存、纯5G手机制造商是利好 $小米集团-W( <span class='highlight'>01810</span> )$ "", 'type': '0', 'source_link': 'http://xueqiu.com/about/mobile-xueqiu', 'edited_at': None, 'pic': '', 'user': {'subscribeable': False, 'remark': None, 'status': 2, 'common_count': 0, 'domain': None, 'description': '如果觉得我说的不对，请用数据和事实来打我脸，而不是想当然和情绪。', 'type': '1', 'location': None, 'id': 4706271602, 'stats': None, 'url': None, 'screen_name': '小米情报官', 'following': False, 'profile': '/4706271602', 'verified': False, 'blocking': False, 'donate_count': 0, 'friends_count': 12, 'followers_count': 118, 'st_color': '1', 'follow_me': False, 'verified_description': None, 'province': '北京', 'city': '不限', 'gender': 'm', 'step': 'null', 'blog_description': None, 'verified_type': 0, 'last_status_id': 142932571, 'status_count': 171, 'recommend': None, 'intro': None, 'stock_status_count': None, 'allow_all_stock': False, 'stocks_count': None, 'name': None, 'group_ids': None, 'name_pinyin': None, 'screenname_pinyin': None, 'block_status': 0, '_source': None, 'verified_realname': False, 'verified_infos': None, 'profile_image_url': 'community/20201/1582703078929-1582703083268.jpg,community/20201/1582703078929-1582703083268.jpg!180x180.png,community/20201/1582703078929-1582703083268.jpg!50x50.png,community/20201/1582703078929-1582703083268.jpg!30x30.png', 'photo_domain': '//xavatar.imedao.com/'}, 'retweeted_status': None, 'target': '/4706271602/142926821', 'fragment': None, 'blocked': False, 'blocking': False, 'topic_pic': None, 'topic_symbol': None, 'topic_title': None, 'topic_desc': None, 'donate_count': 0, 'donate_snowcoin': 0, 'view_count': 0, 'offer': None, 'mark': 0, 'comment': None, 'expend': False, 'rqtype': None, 'rqid': 0, 'firstImg': None, 'card': None, 'favorited': False, 'favorited_created_at': None, 'topic_pic_thumbnail': None, 'topic_pic_headOrPad': None, 'topic_pic_thumbnail_small': None, 'canEdit': True, 'shortStatusMultiPic': False, 'timeBefore': '今天 22:04', 'mark_desc': None, 'reward_count': 0, 'reward_user_count': 0, 'paid_mention': None, 'is_answer': False, 'is_refused': False, 'is_reward': True, '_source': 'sc_1:10:01810', 'pic_sizes': [], 'like_count': 1, 'liked': False, 'promotion_pic': None, 'promotion_url': None, 'promotion_id': 0, 'quote_cards': None, 'is_ss_multi_pic': False, 'excellent_comments': None, 'is_bonus': False, 'source': 'Android'}"
143051676,8436642666,1583423620000,世纪阳光山东基地退城入园估计今年卖地收入60亿不到，另外江西二期85万吨生态肥料产能年中即将投产，今年业绩稳了，明年业绩将高速增长可以熬出头了，现在PE1.8，不可思议。,"{'id': 143051676, 'user_id': 8436642666, 'title': '', 'text': '<a href=""http://xueqiu.com/S/00509"" target=""_blank"">$世纪阳光(00509)$</a> <a href=""http://xueqiu.com/S/00029"" target=""_blank"">$达力集团(00029)$</a> <a href=""http://xueqiu.com/S/01810"" target=""_blank"">$小米集团-W(<span class=\'highlight\'>01810</span>)$</a> 世纪阳光山东基地退城入园估计今年卖地收入60亿不到，另外江西二期85万吨生态肥料产能年中即将投产，今年业绩稳了，明年业绩将高速增长可以熬出头了，现在PE1.8，不可思议。<br/><img class=""ke_img"" src=""https://xqimg.imedao.com/170ab66fafb117d33fea35d0.jpeg!custom.jpg"" /><br/><img class=""ke_img"" src=""https://xqimg.imedao.com/170ab66fff11191a3fe1c64c.jpeg!custom.jpg"" />', 'created_at': 1583423620000, 'retweet_count': 0, 'reply_count': 0, 'fav_count': 0, 'truncated': False, 'commentId': 0, 'retweet_status_id': 0, 'symbol_id': None, 'description': ' <a href=""http://xueqiu.com/S/00509"" target=""_blank"">$世纪阳光(00509)$</a> <a href=""http://xueqiu.com/S/00029"" target=""_blank"">$达力集团(00029)$</a> $小米集团-W( <span class=\'highlight\'>01810</span> )$ 世纪阳光山东基地退城入园估计今年卖地收入60亿不到，另外江西二期85万吨生态肥料产能年中即将投产，今年业绩稳了，明年业绩将高速增长可以熬出头了，现在PE1.8，不可思议。 ', 'type': '2', 'source_link': 'http://xueqiu.com/about/mobile-xueqiu', 'edited_at': None, 'pic': 'https://xqimg.imedao.com/170ab66fafb117d33fea35d0.jpeg!thumb.jpg,https://xqimg.imedao.com/170ab66fff11191a3fe1c64c.jpeg!thumb.jpg', 'user': {'subscribeable': False, 'remark': None, 'status': 1, 'common_count': 0, 'domain': None, 'description': '', 'type': '1', 'location': None, 'id': 8436642666, 'url': None, 'screen_name': '任真惦', 'following': False, 'profile': '/8436642666', 'verified': False, 'blocking': False, 'donate_count': 0, 'friends_count': 210, 'followers_count': 76, 'st_color': '1', 'follow_me': False, 'verified_description': None, 'province': '广东', 'city': '揭阳', 'gender': 'm', 'status_count': 93, 'step': 'two', 'blog_description': None, 'verified_type': 0, 'last_status_id': 143051676, 'stats': None, 'recommend': None, 'intro': None, 'stock_status_count': None, 'allow_all_stock': False, 'stocks_count': None, 'name': None, 'group_ids': None, 'name_pinyin': None, 'screenname_pinyin': None, 'block_status': 0, '_source': None, 'verified_realname': False, 'verified_infos': None, 'profile_image_url': 'community/20183/1524671516633-1524671516796.jpeg,community/20183/1524671516633-1524671516796.jpeg!180x180.png,community/20183/1524671516633-1524671516796.jpeg!50x50.png,community/20183/1524671516633-1524671516796.jpeg!30x30.png', 'photo_domain': '//xavatar.imedao.com/'}, 'retweeted_status': None, 'target': '/8436642666/143051676', 'fragment': None, 'blocked': False, 'blocking': False, 'topic_pic': None, 'topic_symbol': None, 'topic_title': None, 'topic_desc': None, 'donate_count': 0, 'donate_snowcoin': 0, 'view_count': 0, 'offer': None, 'mark': 0, 'comment': None, 'expend': True, 'rqtype': None, 'rqid': 0, 'firstImg': 'https://xqimg.imedao.com/170ab66fafb117d33fea35d0.jpeg!thumb.jpg', 'card': None, 'favorited': False, 'favorited_created_at': None, 'topic_pic_thumbnail': None, 'topic_pic_headOrPad': None, 'topic_pic_thumbnail_small': None, 'canEdit': True, 'shortStatusMultiPic': True, 'timeBefore': '7分钟前', 'mark_desc': None, 'reward_count': 0, 'reward_user_count': 0, 'paid_mention': None, 'is_answer': False, 'is_refused': False, 'is_reward': True, '_source': 'sc_1:1:01810', 'pic_sizes': [{'width': 1080, 'height': 1920, 'ratio': '0.56', 'score': 3, 'setScore': True, 'setRatio': True, 'setWidth': True, 'setHeight': True}, {'width': 1080, 'height': 1920, 'ratio': '0.56', 'score': 1, 'setScore': True, 'setRatio': True, 'setWidth': True, 'setHeight': True}], 'like_count': 0, 'liked': False, 'promotion_pic': None, 'promotion_url': None, 'promotion_id': 0, 'quote_cards': None, 'is_ss_multi_pic': True, 'excellent_comments': None, 'source': 'Android', 'is_bonus': False}"
143051525,4281679676,1583423489000,#facebook# ，马云联手100家银行，在金融圈引发震动！马云那句银行不改变，我们就改变银行，曾经深深影响了银行圈。很多人认为马云要颠覆银行。多年过去，今天回头看这句话，多少有些被曲解！先来看下，马云这次和100家银行联手，是要做什么？根据公开信息，这次是马云旗下的网商银行，要和100家银行联手大推「无接触贷款」，而且这件事，国家非常支持，全国工商联亲自挂帅，作为计划的发起方。其实从这件事的本质来看，再一次打脸传统银行，再一次引领银行业的未来发展和方向！在传统经济和互联网经济的结合，其本质还是小米的互联网手机思路，是互联网工具的用户体验模式，所有生意的本质都是流量和用户体验而带来的口碑模式，未来的发展是用户的时代，是用户至上的口碑时代，是用户体验到客户价值导向！未来银行业的发展，谁会好过，谁会过得滋润，谁会更艰难，这是一个互联网引领的银行裂变时代，是一个撕裂银行的时代，有的银行注定粉身碎骨，被历史的车轮碾压得面目全非，而有的银行将是鲜花和掌声包围！未来银行一定是存在的，但会是以什么方式存在以及成为霸主，但一定是跟消费，信贷，支付，数字货币，互联网融合的混业经营以及糅合的银行，我们是拭目以待！所以，投资银行股风险是巨大的，不要被低价和打折迷惑了你购买标的本质，在国内投资A股不要被股息和低市盈率估值蒙蔽，银行股不要轻信投资专家的分析，投资一定要看企业的持续盈利能力而不是短期盈利，也就是永续经营的能力，护城河的能力，企业文化导向的能力！银行业未来真正的霸主强者，未必是银行主导，未来一切都难说，本人一直看好互联网导入的控股和组建的银行金融业态，比如像小米金融，腾讯和阿里以及京东的这类金融，当然，但一直以来我仅仅代表我个人的观点！所以选择投资标的，还是要以价值投资和永续经营为前提！未来，顺丰有自己或者自己控股的银行，Facebook目前在布局的棋比腾讯还大，苹果和亚马逊也是！小米有自己或者自己控股的银行，其实小米已经在做了，做得还不错，未来在中国，更看好小米的互联网思维做混业经营。但由于茅台是国企，所以很多性质不同，其实茅台财务公司已经在干这个事，只是没有巴菲特段永平这种能力而已！总而言之：能够颠覆银行的，一定不是银行，而是跨界！,"{'id': 143051525, 'user_id': 4281679676, 'title': '', 'text': '<a href=""http://xueqiu.com/S/01810"" target=""_blank"">$小米集团-W(<span class=\'highlight\'>01810</span>)$</a> <a href=""http://xueqiu.com/S/SH600519"" target=""_blank"">$贵州茅台(SH600519)$</a> <a href=""http://xueqiu.com/k?q=%23facebook%23"" target=""_blank"">#facebook#</a> ，马云联手100家银行，在金融圈引发震动！马云那句银行不改变，我们就改变银行，曾经深深影响了银行圈。很多人认为马云要颠覆银行。<br/>多年过去，今天回头看这句话，多少有些被曲解！<br/>先来看下，马云这次和100家银行联手，是要做什么？根据公开信息，这次是马云旗下的网商银行，要和100家银行联手大推「无接触贷款」，而且这件事，国家非常支持，全国工商联亲自挂帅，作为计划的发起方。<br/>其实从这件事的本质来看，再一次打脸传统银行，再一次引领银行业的未来发展和方向！在传统经济和互联网经济的结合，其本质还是小米的互联网手机思路，是互联网工具的用户体验模式，所有生意的本质都是流量和用户体验而带来的口碑模式，未来的发展是用户的时代，是用户至上的口碑时代，是用户体验到客户价值导向！未来银行业的发展，谁会好过，谁会过得滋润，谁会更艰难，这是一个互联网引领的银行裂变时代，是一个撕裂银行的时代，有的银行注定粉身碎骨，被历史的车轮碾压得面目全非，而有的银行将是鲜花和掌声包围！未来银行一定是存在的，但会是以什么方式存在以及成为霸主，但一定是跟消费，信贷，支付，数字货币，互联网融合的混业经营以及糅合的银行，我们是拭目以待！所以，投资银行股风险是巨大的，不要被低价和打折迷惑了你购买标的本质，在国内投资A股不要被股息和低市盈率估值蒙蔽，银行股不要轻信投资专家的分析，投资一定要看企业的持续盈利能力而不是短期盈利，也就是永续经营的能力，护城河的能力，企业文化导向的能力！银行业未来真正的霸主强者，未必是银行主导，未来一切都难说，本人一直看好互联网导入的控股和组建的银行金融业态，比如像小米金融，腾讯和阿里以及京东的这类金融，当然，但一直以来我仅仅代表我个人的观点！所以选择投资标的，还是要以价值投资和永续经营为前提！未来，顺丰有自己或者自己控股的银行，Facebook目前在布局的棋比腾讯还大，苹果和亚马逊也是！小米有自己或者自己控股的银行，其实小米已经在做了，做得还不错，未来在中国，更看好小米的互联网思维做混业经营。但由于茅台是国企，所以很多性质不同，其实茅台财务公司已经在干这个事，只是没有巴菲特段永平这种能力而已！总而言之：能够颠覆银行的，一定不是银行，而是跨界！', 'created_at': 1583423489000, 'retweet_count': 0, 'reply_count': 0, 'fav_count': 0, 'truncated': False, 'commentId': 0, 'retweet_status_id': 0, 'symbol_id': None, 'description': ' $小米集团-W( <span class=\'highlight\'>01810</span> )$ <a href=""http://xueqiu.com/S/SH600519"" target=""_blank"">$贵州茅台(SH600519)$</a> <a href=""http://xueqiu.com/k?q=%23facebook%23"" target=""_blank"">#facebook#</a> ，马云联手100家银行，在金融圈引发震动！马云那句银行不改变，我们就改变银行，曾经深深影响了银行圈。很多人认为马云要颠覆银行。 多年过去，今天回头看这句话，多少有些被曲解！ 先来看下，马云这次和100家银行联手，是要做什么？根据公开信息，...', 'type': '2', 'source_link': 'http://xueqiu.com/about/mobile-xueqiu', 'edited_at': None, 'pic': '', 'user': {'subscribeable': False, 'remark': None, 'status': 2, 'common_count': 0, 'domain': None, 'description': '有为，有不为', 'type': '1', 'location': None, 'id': 4281679676, 'url': None, 'screen_name': '孤独的狂欢蜗牛', 'following': False, 'profile': '/4281679676', 'verified': False, 'blocking': False, 'donate_count': 0, 'friends_count': 221, 'followers_count': 577, 'st_color': '1', 'follow_me': False, 'verified_description': None, 'province': '海外', 'city': '新加坡', 'gender': 'm', 'status_count': 465, 'step': 'null', 'blog_description': None, 'verified_type': 0, 'last_status_id': 143051525, 'stats': None, 'recommend': None, 'intro': None, 'stock_status_count': None, 'allow_all_stock': False, 'stocks_count': None, 'name': None, 'group_ids': None, 'name_pinyin': None, 'screenname_pinyin': None, 'block_status': 0, '_source': None, 'verified_realname': False, 'verified_infos': None, 'profile_image_url': 'community/20197/1564594898226-1564594898504.jpeg,community/20197/1564594898226-1564594898504.jpeg!180x180.png,community/20197/1564594898226-1564594898504.jpeg!50x50.png,community/20197/1564594898226-1564594898504.jpeg!30x30.png', 'photo_domain': '//xavatar.imedao.com/'}, 'retweeted_status': None, 'target': '/4281679676/143051525', 'fragment': None, 'blocked': False, 'blocking': False, 'topic_pic': None, 'topic_symbol': None, 'topic_title': None, 'topic_desc': None, 'donate_count': 0, 'donate_snowcoin': 0, 'view_count': 0, 'offer': None, 'mark': 0, 'comment': None, 'expend': True, 'rqtype': None, 'rqid': 0, 'firstImg': None, 'card': {'param': 'typ=MANUAL&id=299', 'data': '{""items"":[{""id"":299,""tag"":""#facebook#"",""content"":"""",""pic"":null,""backgroundColor"":"""",""following"":false,""pic_size"":null,""status_count"":36,""follower_count"":0}]}', 'first': '{""items"":[{""id"":299,""tag"":""#facebook#"",""content"":"""",""pic"":null,""backgroundColor"":"""",""following"":false,""pic_size"":null,""status_count"":36,""follower_count"":0}]}', 'type': 'hashtag'}, 'favorited': False, 'favorited_created_at': None, 'topic_pic_thumbnail': None, 'topic_pic_headOrPad': None, 'topic_pic_thumbnail_small': None, 'canEdit': True, 'shortStatusMultiPic': False, 'timeBefore': '9分钟前', 'mark_desc': None, 'reward_count': 0, 'reward_user_count': 0, 'paid_mention': None, 'is_answer': False, 'is_refused': False, 'is_reward': True, '_source': 'sc_1:2:01810', 'pic_sizes': [], 'like_count': 0, 'liked': False, 'promotion_pic': None, 'promotion_url': None, 'promotion_id': 0, 'quote_cards': None, 'is_ss_multi_pic': False, 'excellent_comments': None, 'source': 'Android', 'is_bonus': False}"
143051046,4104510874,1583423092000,&nbsp; 根据雷军在2018年提出的目标：10年营收增长10倍。跟这个算法，小米的年均要接近25%的高速增长，10年后才能达到销售额17000亿10年增长10倍，A股市场不乏许多经典案例，比方格力电器，2002年销售额70亿，到了2012年销售额1001亿，增长14倍恒瑞医药，2002年销售额7.5亿，一直到2014年才达到10倍左右的74亿腾讯，2007年销售额38亿，2017年2377亿，接近62倍当然了，上面的对比都太简单了我想说的是，市场没有天花板，修行在各人可能大家都在想，小米体量太大，没可能华为2019年销售收入8500亿，10年后，华为会达到8万5000亿的销售额吗》？如果小米在十年后，取代OV，一统物联网江湖，成为国内第二，国际前三能够达成年度销售收入17000亿的目标吗？印钞的货币是越来越多的目前来看，小米达成目标有三成的机会大家觉得，会有几成胜算？（如果达成10年10倍的目标，股价会增长几倍，还是有点小期待）,"{'id': 143051046, 'user_id': 4104510874, 'title': '', 'text': '<p><a href=""http://xueqiu.com/S/01810"" target=""_blank"">$小米集团-W(<span class=\'highlight\'>01810</span>)$</a>&nbsp;<a href=""http://xueqiu.com/S/SZ000651"" target=""_blank"">$格力电器(SZ000651)$</a> 根据雷军在2018年提出的目标：10年营收增长10倍。</p><p>跟这个算法，小米的年均要接近25%的高速增长，10年后才能达到销售额17000亿</p><p>10年增长10倍，A股市场不乏许多经典案例，</p><p>比方格力电器，2002年销售额70亿，到了2012年销售额1001亿，增长14倍</p><p>恒瑞医药，2002年销售额7.5亿，一直到2014年才达到10倍左右的74亿</p><p>腾讯，2007年销售额38亿，2017年2377亿，接近62倍</p><p>当然了，上面的对比都太简单了</p><p>我想说的是，市场没有天花板，修行在各人</p><p>可能大家都在想，小米体量太大，没可能</p><p>华为2019年销售收入8500亿，10年后，华为会达到8万5000亿的销售额吗》？</p><p>如果小米在十年后，取代OV，一统物联网江湖，成为国内第二，国际前三</p><p>能够达成年度销售收入17000亿的目标吗？</p><p>印钞的货币是越来越多的</p><p>目前来看，小米达成目标有三成的机会</p><p>大家觉得，会有几成胜算？</p><p>（如果达成10年10倍的目标，股价会增长几倍，还是有点小期待）</p>', 'created_at': 1583423092000, 'retweet_count': 0, 'reply_count': 0, 'fav_count': 0, 'truncated': False, 'commentId': 0, 'retweet_status_id': 0, 'symbol_id': None, 'description': ' $小米集团-W( <span class=\'highlight\'>01810</span> )$   <a href=""http://xueqiu.com/S/SZ000651"" target=""_blank"">$格力电器(SZ000651)$</a> 根据雷军在2018年提出的目标：10年营收增长10倍。 跟这个算法，小米的年均要接近25%的高速增长，10年后才能达到销售额17000亿 10年增长10倍，A股市场不乏许多经典案例， 比方格力电器，2002年销售额70亿，到了2012年销售额1001亿，增长14倍 恒瑞医药，2002年...', 'type': '2', 'source_link': None, 'edited_at': None, 'pic': '', 'user': {'subscribeable': False, 'remark': None, 'status': 2, 'common_count': 0, 'domain': None, 'description': '熊市中防御性坚守，不出局你就赢了\n喜欢的一起唠唠，V：pande633', 'type': '1', 'location': None, 'id': 4104510874, 'url': None, 'screen_name': '华山派大师哥', 'following': False, 'profile': '/4104510874', 'verified': False, 'blocking': False, 'donate_count': 0, 'friends_count': 69, 'followers_count': 169, 'st_color': '1', 'follow_me': False, 'verified_description': '', 'province': '江苏', 'city': '徐州', 'gender': 'm', 'status_count': 363, 'step': 'null', 'blog_description': None, 'verified_type': 0, 'last_status_id': 143051046, 'stats': None, 'recommend': None, 'intro': None, 'stock_status_count': None, 'allow_all_stock': False, 'stocks_count': None, 'name': None, 'group_ids': None, 'name_pinyin': None, 'screenname_pinyin': None, 'block_status': 0, '_source': None, 'verified_realname': False, 'verified_infos': [{'verified_type': '5', 'verified_desc': '用户已完成实名身份认证'}], 'profile_image_url': 'community/201810/1543331700605-1543331700791.jpeg,community/201810/1543331700605-1543331700791.jpeg!180x180.png,community/201810/1543331700605-1543331700791.jpeg!50x50.png,community/201810/1543331700605-1543331700791.jpeg!30x30.png', 'photo_domain': '//xavatar.imedao.com/'}, 'retweeted_status': None, 'target': '/4104510874/143051046', 'fragment': None, 'blocked': False, 'blocking': False, 'topic_pic': None, 'topic_symbol': None, 'topic_title': None, 'topic_desc': None, 'donate_count': 0, 'donate_snowcoin': 0, 'view_count': 0, 'offer': None, 'mark': 0, 'comment': None, 'expend': True, 'rqtype': None, 'rqid': 0, 'firstImg': None, 'card': None, 'favorited': False, 'favorited_created_at': None, 'topic_pic_thumbnail': None, 'topic_pic_headOrPad': None, 'topic_pic_thumbnail_small': None, 'canEdit': True, 'shortStatusMultiPic': False, 'timeBefore': '16分钟前', 'mark_desc': None, 'reward_count': 0, 'reward_user_count': 0, 'paid_mention': None, 'is_answer': False, 'is_refused': False, 'is_reward': True, '_source': 'sc_1:3:01810', 'pic_sizes': [], 'like_count': 1, 'liked': False, 'promotion_pic': None, 'promotion_url': None, 'promotion_id': 0, 'quote_cards': None, 'is_ss_multi_pic': False, 'excellent_comments': None, 'source': '雪球', 'is_bonus': False}"
143050231,7889792053,1583422557000,,"{'id': 143050231, 'user_id': 7889792053, 'title': '', 'text': '<a href=""http://xueqiu.com/S/01810"" target=""_blank"">$小米集团-W(<span class=\'highlight\'>01810</span>)$</a> <img src=""//assets.imedao.com/ugc/images/face/emoji_02_laughing.png"" title=""[大笑]"" alt=""[大笑]"" height=""24"" />', 'created_at': 1583422557000, 'retweet_count': 0, 'reply_count': 0, 'fav_count': 0, 'truncated': False, 'commentId': 139505275, 'retweet_status_id': 142940082, 'symbol_id': None, 'description': "" $小米集团-W( <span class='highlight'>01810</span> )$ "", 'type': None, 'source_link': 'http://xueqiu.com/about/mobile-xueqiu', 'edited_at': None, 'pic': '', 'user': {'subscribeable': False, 'remark': None, 'status': 0, 'common_count': 0, 'domain': None, 'description': '', 'type': '1', 'location': None, 'id': 7889792053, 'url': None, 'screen_name': 'hi大魔王', 'following': False, 'profile': '/7889792053', 'verified': False, 'blocking': False, 'donate_count': 0, 'friends_count': 18, 'followers_count': 1, 'st_color': '1', 'follow_me': False, 'verified_description': None, 'province': '不限', 'city': '未知', 'gender': 'n', 'status_count': 14, 'step': 'null', 'blog_description': None, 'verified_type': 0, 'last_status_id': 143052109, 'stats': None, 'recommend': None, 'intro': None, 'stock_status_count': None, 'allow_all_stock': False, 'stocks_count': None, 'name': None, 'group_ids': None, 'name_pinyin': None, 'screenname_pinyin': None, 'block_status': 0, '_source': None, 'verified_realname': False, 'verified_infos': None, 'profile_image_url': 'community/default/avatar.png,community/default/avatar.png!180x180.png,community/default/avatar.png!50x50.png,community/default/avatar.png!30x30.png', 'photo_domain': '//xavatar.imedao.com/'}, 'retweeted_status': {'id': 142940082, 'user_id': 1713446664, 'title': '小米倒闭？华为余承东五年前预测落空 雷军也没能让投资者挣一倍', 'text': '<p><img class=""ke_img"" src=""https://xqimg.imedao.com/170a6722660f90e3fc682794.jpg!custom.jpg"" /></p><p><b>雷达财经出品 文|长帆 编|深海</b></p><p>“要买小米的抓紧啦，小米还有一天就倒闭了。”近两日，余承东五年前讲话被网友翻出。2015年3月4日，腾讯科技报道称，余承东预测未来三到五年，小米将要消失。网友据此创作了多个段子。</p><p>雷军表示，小米在微博上已经“被关门”很多次了。3月2日，小米中国区总裁卢伟冰微博上转发了雷军的微博，称还有两天就“关门”了，疑似对余承东预言的隔空回应。</p><p><img class=""ke_img"" src=""https://xqimg.imedao.com/170a672269bf9613fbf40c3f.jpg!custom.jpg"" /></p><p>截至2020年3月4日，余承东预言小米消失已满5年，小米依然存活。截至3月4日收盘，小米市值为3107.2亿港元。</p><p><img class=""ke_img"" src=""https://xqimg.imedao.com/170a6722670fae03fe735fd2.jpg!custom.jpg"" /></p><p>余承东在手机界有“余大嘴”之称。相比之下，小米的“大嘴”更多，除了雷军本人，小米还将卢伟冰、常程揽入麾下。有传言称，罗永浩也可能加入小米。</p><p>作为小米创始人，雷军也曾立下多个“flag”。如雷军在上市庆功宴上称，让上市首日购买小米的投资者挣一倍。</p><p>雷达财经统计发现，自小米2018年7月9日上市至今，股价累计下跌达23.88%。这意味着，响应雷军号召购买股票的投资者，平均亏损超两成。</p><p><br/></p><p><b>五年期满小米市值3100亿港元</b></p><p>2015年3月4日腾讯科技一则新闻，标题为：“华为余承东：5年内小米将消失”。</p><p></p><p>当时，余承东在世界移动通信大会上发表言论，智能手机行业将重新洗牌，未来三到五年只会剩下三大手机厂商。</p><p>余承东认为华为将是三大手机厂商之一。但他承认，对于整个行业来说，这将是痛苦的，现今的行业领军者可能会被迅速遗忘。</p><p>“未来，我们的主要对手将是西方公司，而不是中国公司。未来三到五年多数中国智能手机厂商将消失，我们将成为市场领军者。”余承东说道。</p><p>谈到小米，余承东表示：“很多手机厂商看起来非常成功，但它们将很快消失。从长远来看，我认为它们无法生存。”</p><p>随着五年之期临近，多位网友翻出了腾讯当时的报道，并发起微博话题<a href=""http://xueqiu.com/k?q=%23%E6%98%8E%E5%A4%A9%E5%B0%8F%E7%B1%B3%E5%B0%B1%E8%A6%81%E5%80%92%E9%97%AD%E4%BA%86%23"" target=""_blank"">#明天小米就要倒闭了#</a>。截至3月4日晚十点，该话题阅读量389.4万。</p><p>截至3月4日收盘，小米报收于12.94港元/股，上涨1.25%，市值3107.2亿港元。</p><p>“我帮余承东想好了一个下台阶的好办法，小米现在市值400亿美金，华为拿出200亿美金控股小米，然后明天宣布小米解散，这样余承东吹过的牛就可以实现了。”有网友在3月3日给余承东支招。</p><p>雷达财经梳理发现，余承东在手机圈有“大嘴”之称。其最爱怼的是苹果，曾称“带充电宝的一般都是iPhone用户”。</p><p>作为华为国内对手，小米也没少遭到余承东的“炮轰”。2014年，网上有一篇评测文章认为，荣耀6比小米4更好，雷军询问“这是哪位友商的杰作？”余承东则回应称：“询问了内部团队没有这样的事，建议雷总大气一点。”</p><p><br/></p><p><b>小米吸引多位手机圈“大嘴”加盟</b></p><p>对于余承东的“炮轰”，小米毫不示弱。</p><p>2019年9月，华为上海召开新品发布会，推出了Mate 30系列旗舰机、华为智慧屏等产品。发布会后的采访中，余承东向媒体表示，华为的5G手机比对手领先半年到一年以上。此外，他还不忘怼友商：“（小米MIX Alpha）只是概念手机，没有实用价值。”</p><p>对此，卢伟冰微博上公开呛声余承东：“如果说环绕屏没有‘实用价值’，那么把一个外国车牌贴在手机上就多卖1万块的‘实用价值’是什么？”卢伟冰直指华为Mate 30 RS保时捷设计手机没有“实用价值”。</p><p>卢伟冰在手机圈也有“大嘴”之称。其在5G手机、市场排名、库存等问题上，频频向华为荣耀“开火”。</p><p>卢伟冰曾出任金立集团副总裁。2019年1月2日，小米创始人雷军正式宣布，卢伟冰加盟小米。</p><p>小米吸引“大嘴”之路并未到此终止。</p><p>2020年1月2日，雷军在个人微博上宣布联想集团原副总裁常程加入小米任集团副总裁，负责手机产品规划。随后，常程在个人微博转发确定了加盟小米的消息。</p><p>常程，外号“掌柜”，在微博上几乎天天回复粉丝“早上好”，微博粉丝超300万。与由于经常“碰瓷”蹭热点，常程也被人称为“万瓷王”。</p><p>小米曾是常程吐槽对象之一。如2018年10月，为了给即将发布的联想S5 Pro造势，常程在微博吐槽小米8青春版没有无损光学变焦，何谈青春无敌；再是没有前置2000万双摄，何谈微整形美颜。</p><p>罗永浩也是手机界知名“嘴炮”，其在做锤子手机时，每次新品发布会都仿佛相声大会。其知名言论包括：</p><p>“看起来苹果要完蛋了。” “雷军和黄章从来就是土包子。” “你们这些人就不用担心了，我们会全面超越苹果。” </p><p>值得一提的是，今年3 月 2 日下午，罗永浩在微博上连续转发了数条关于黑鲨品牌即将推出的新款手机黑鲨 3 Pro 的相关内容。随后，雷军、卢伟冰、常程、王腾等多位小米管理层人员也转发了这条微博，引发大量网友关注。</p><p><img class=""ke_img"" src=""https://xqimg.imedao.com/170a672266dfac33fe0d5d1f.jpg!custom.jpg"" /></p><p>有网友猜测，罗永浩可能也要入职小米。</p><p>罗永浩曾对小米模式颇为赞赏。2012 年 10 月，罗永浩在一场演讲中表示，“当众我老夸小米他那个 MIUI 做得很好，是定制版的安卓里面做得最好的，但是小米公司最大的问题就是不会做减法。”</p><p>罗永浩自己做手机后，没少怼小米。罗永浩曾说，苹果的竞争对手是一群笨蛋——当然，这群笨蛋中也包括了小米。</p><p><br/></p><p><b>雷军没能让投资者赚一倍</b></p><p>作为小米的掌门人，雷军在怼“友商”这件事上也一直亲力亲为。其怼过的名单很长，包括周鸿祎、黄章、常程、余承东……</p><p>红米Note2推出时，雷军曾经发博称“为何友商很反感跑分？”配图显示，红米Note2跑分完胜当时的华为P8和魅族MX5。魅族掌门人黄章回击称：“只懂得看跑分的是**丝”。</p><p>雷军最为外界津津乐道的是和董明珠的赌局。</p><p>2013年12月12日，在央视财经频道主办的第十四届中国经济年度人物颁奖盛典上，雷军与董明珠约定“10亿赌局”，比5年后公司营收。</p><p>雷军更请全国人民作证，5年内如果小米营业额击败格力，只要董明珠输1块钱就行了。董明珠立即回应道：“第一，我告诉你不可能；第二，要赌不是1块钱，我跟你赌10个亿。”格力和小米的10亿赌约就此展开。</p><p>董明珠和雷军的赌局，成为财经热门话题，多次被媒体提及。</p><p>根据格力电器披露的2018年年报显示，公司营业总收入2000.24亿元、年增率33.33%，而小米2018年总营收为1749亿元。董明珠赢得了10亿赌局。不过，董明珠未找雷军收钱。</p><p>在和董明珠豪赌10亿期间，雷军带领小米在香港上市。2018年7月9日，雷军在上市敲钟的庆功宴上承诺说，“要让上市首日买入小米公司股票的投资人赚一倍。”</p><p>不过，首日买入小米股票的投资者不仅未能挣到1倍，反而出现了亏损。雷达财经统计发现，小米上市至今，股价累计下跌达23.88%。</p><p><img class=""ke_img"" src=""https://xqimg.imedao.com/170a67226c0f90f3febe85d1.jpg!custom.jpg"" /></p><p>财务数据显示，小米收入出现放缓迹象。2019年第三季度，小米公司收入为人民币537亿元，较第二季度报告的人民币519.5亿元，增长3.3％。而小米第二季度环比增长18.7％，同比增长14.8％。</p><p>第三季度小米手机出货量3210万部，同比下滑3.6％，收入323亿元人民币，同比下滑7.8％。</p><p>随着新冠肺炎疫情爆发，手机行业也受到巨大冲击。2020年2月，卢伟冰在微博上透露，中国手机市场1月大约3000万部，2月由于疫情原因，目前看市场大盘预计在1500万以下，整个市场环比下滑50%以上。不过，卢伟冰称，小米受影响最小。</p><p><img class=""ke_img"" src=""https://xqimg.imedao.com/170a67228e4fba13fdab46c8.jpg!custom.jpg"" /></p><p>小米能否摆脱增速下滑？疫情冲击下小米能否弯道超车？小米的投资者何时能解套？雷达财经将继续关注。</p><p><i>注：本文是雷达财经（ID：leidacj）原创。未经授权，禁止转载。</i></p><p><a href=""http://xueqiu.com/S/01810"" target=""_blank"">$小米集团-W(01810)$</a> <a href=""http://xueqiu.com/S/00992"" target=""_blank"">$联想集团(00992)$</a> <a href=""http://xueqiu.com/S/SZ000651"" target=""_blank"">$格力电器(SZ000651)$</a> </p>', 'created_at': 1583340552000, 'retweet_count': 2, 'reply_count': 3, 'fav_count': 0, 'truncated': False, 'commentId': 0, 'retweet_status_id': 0, 'symbol_id': None, 'description': '雷达财经出品 文|长帆 编|深海 “要买小米的抓紧啦，小米还有一天就倒闭了。”近两日，余承东五年前讲话被网友翻出。2015年3月4日，腾讯科技报道称，余承东预测未来三到五年，小米将要消失。网友据此创作了多个段子。 雷军表示，小米在微博上已经“被关门”很多次了。3月2日，小米中国区总裁卢伟...', 'type': '3', 'source_link': None, 'edited_at': None, 'pic': 'https://xqimg.imedao.com/170a6722660f90e3fc682794.jpg!thumb.jpg,https://xqimg.imedao.com/170a672269bf9613fbf40c3f.jpg!thumb.jpg,https://xqimg.imedao.com/170a6722670fae03fe735fd2.jpg!thumb.jpg,https://xqimg.imedao.com/170a672266dfac33fe0d5d1f.jpg!thumb.jpg,https://xqimg.imedao.com/170a67226c0f90f3febe85d1.jpg!thumb.jpg,https://xqimg.imedao.com/170a67228e4fba13fdab46c8.jpg!thumb.jpg', 'user': {'subscribeable': False, 'remark': None, 'status': 1, 'common_count': 0, 'domain': None, 'description': '把脉公司全方位，扫除雷点，发现价值！', 'type': '1', 'location': None, 'id': 1713446664, 'url': None, 'screen_name': '雷达财经', 'following': False, 'profile': '/1713446664', 'verified': False, 'blocking': False, 'donate_count': 0, 'friends_count': 36, 'followers_count': 80, 'st_color': '1', 'follow_me': False, 'verified_description': None, 'province': '北京', 'city': '海淀区', 'gender': 'm', 'status_count': 173, 'step': 'two', 'blog_description': None, 'verified_type': 0, 'last_status_id': 142940082, 'stats': None, 'recommend': None, 'intro': None, 'stock_status_count': None, 'allow_all_stock': False, 'stocks_count': None, 'name': None, 'group_ids': None, 'name_pinyin': None, 'screenname_pinyin': None, 'block_status': 0, '_source': None, 'verified_realname': False, 'verified_infos': None, 'profile_image_url': 'community/20199/1571148585525-1571148589997.jpeg,community/20199/1571148585525-1571148589997.jpeg!180x180.png,community/20199/1571148585525-1571148589997.jpeg!50x50.png,community/20199/1571148585525-1571148589997.jpeg!30x30.png', 'photo_domain': '//xavatar.imedao.com/'}, 'retweeted_status': None, 'target': '/1713446664/142940082', 'fragment': None, 'blocked': False, 'blocking': False, 'topic_pic': None, 'topic_symbol': None, 'topic_title': None, 'topic_desc': None, 'donate_count': 0, 'donate_snowcoin': 0, 'view_count': 21646, 'offer': None, 'mark': 0, 'comment': None, 'expend': True, 'rqtype': None, 'rqid': 0, 'firstImg': 'https://xqimg.imedao.com/170a6722660f90e3fc682794.jpg!thumb.jpg', 'card': {'param': 'typ=MANUAL&id=103953', 'data': '{""items"":[{""id"":103953,""tag"":""#明天小米就要倒闭了#"",""content"":"""",""pic"":null,""backgroundColor"":"""",""following"":false,""pic_size"":null,""status_count"":1,""follower_count"":0}]}', 'first': '{""items"":[{""id"":103953,""tag"":""#明天小米就要倒闭了#"",""content"":"""",""pic"":null,""backgroundColor"":"""",""following"":false,""pic_size"":null,""status_count"":1,""follower_count"":0}]}', 'type': 'hashtag'}, 'favorited': False, 'favorited_created_at': None, 'topic_pic_thumbnail': None, 'topic_pic_headOrPad': None, 'topic_pic_thumbnail_small': None, 'canEdit': True, 'shortStatusMultiPic': False, 'timeBefore': '03-05 00:49', 'mark_desc': None, 'reward_count': 0, 'reward_user_count': 0, 'paid_mention': None, 'is_answer': False, 'is_refused': False, 'is_reward': True, '_source': None, 'pic_sizes': [{'width': 600, 'height': 337, 'ratio': '1.78', 'score': 2, 'setScore': True, 'setRatio': True, 'setWidth': True, 'setHeight': True}, {'width': 640, 'height': 804, 'ratio': '0.80', 'score': 1, 'setScore': True, 'setRatio': True, 'setWidth': True, 'setHeight': True}, {'width': 640, 'height': 503, 'ratio': '1.27', 'score': 1, 'setScore': True, 'setRatio': True, 'setWidth': True, 'setHeight': True}, {'width': 600, 'height': 548, 'ratio': '1.09', 'score': 1, 'setScore': True, 'setRatio': True, 'setWidth': True, 'setHeight': True}, {'width': 291, 'height': 185, 'ratio': '1.57', 'score': 5, 'setScore': True, 'setRatio': True, 'setWidth': True, 'setHeight': True}, {'width': 617, 'height': 436, 'ratio': '1.42', 'score': 3, 'setScore': True, 'setRatio': True, 'setWidth': True, 'setHeight': True}], 'like_count': 1, 'liked': False, 'promotion_pic': None, 'promotion_url': None, 'promotion_id': 0, 'quote_cards': None, 'is_ss_multi_pic': False, 'excellent_comments': None, 'source': '雪球', 'is_bonus': False}, 'target': '/7889792053/143050231', 'fragment': None, 'blocked': False, 'blocking': False, 'topic_pic': None, 'topic_symbol': None, 'topic_title': None, 'topic_desc': None, 'donate_count': 0, 'donate_snowcoin': 0, 'view_count': 0, 'offer': None, 'mark': 0, 'comment': None, 'expend': False, 'rqtype': None, 'rqid': 0, 'firstImg': None, 'card': None, 'favorited': False, 'favorited_created_at': None, 'topic_pic_thumbnail': None, 'topic_pic_headOrPad': None, 'topic_pic_thumbnail_small': None, 'canEdit': False, 'shortStatusMultiPic': False, 'timeBefore': '25分钟前', 'mark_desc': None, 'reward_count': 0, 'reward_user_count': 0, 'paid_mention': None, 'is_answer': False, 'is_refused': False, 'is_reward': True, '_source': 'sc_1:4:01810', 'pic_sizes': [], 'like_count': 0, 'liked': False, 'promotion_pic': None, 'promotion_url': None, 'promotion_id': 0, 'quote_cards': None, 'is_ss_multi_pic': False, 'excellent_comments': None, 'source': 'Android', 'is_bonus': False}"
143048629,2269671151,1583421503000,离开我海半年了，今天写券商文对比，提到我海，然后回来看看。翻看文章，这半年，并没有啥很有价值的文章跟高水平的球友出现，跟我走的时候差不多熊样。我第一次建仓国海是201704，6.48，到最低点2.75（配股除权前）跌了58%，中间不断的跌了加/加了跌，201810在跌停板两块九上买了最大的单笔，首融。然后国海当了一次大哥，有了盈利，后来在201910带着不大的盈利换仓到东财。整个2018年里，我是雪球国海讨论区热度前三吧，我不认为市场里比我拿的久想的多说的多的球友有多少。券商里买国海，前期是赌反转，后来的都是因为那段大哥效应吧？但这个并不能长久。国海是我最痛恨也曾是最欣喜的品种，但过去了，以后还是找确定性，不去赌。以此悼念我当年每天3.33挂国海两个多月，始终差一两分不成交。好处是锻炼了我的耐性。即便这样，我还是称之为我海。就这些。,"{'id': 143048629, 'user_id': 2269671151, 'title': '', 'text': '<a href=""http://xueqiu.com/S/SZ000750"" target=""_blank"">$国海证券(SZ000750)$</a> <br/>离开我海半年了，今天写券商文对比，提到我海，然后回来看看。<br/>翻看文章，这半年，并没有啥很有价值的文章跟高水平的球友出现，跟我走的时候差不多熊样。<br/>我第一次建仓国海是201704，6.48，到最低点2.75（配股除权前）跌了58%，中间不断的跌了加/加了跌，201810在跌停板两块九上买了最大的单笔，首融。然后国海当了一次大哥，有了盈利，后来在201910带着不大的盈利换仓到东财。<br/>整个2018年里，我是雪球国海讨论区热度前三吧，我不认为市场里比我拿的久想的多说的多的球友有多少。券商里买国海，前期是赌反转，后来的都是因为那段大哥效应吧？但这个并不能长久。<br/>国海是我最痛恨也曾是最欣喜的品种，但过去了，以后还是找确定性，不去赌。<br/>以此悼念我当年每天3.33挂国海两个多月，始终差一两分不成交。好处是锻炼了我的耐性。<br/>即便这样，我还是称之为我海。<br/>就这些。', 'created_at': 1583421503000, 'retweet_count': 0, 'reply_count': 6, 'fav_count': 0, 'truncated': False, 'commentId': 0, 'retweet_status_id': 0, 'symbol_id': None, 'description': ' <a href=""http://xueqiu.com/S/SZ000750"" target=""_blank"">$国海证券(SZ000750)$</a> 离开我海半年了，今天写券商文对比，提到我海，然后回来看看。 翻看文章，这半年，并没有啥很有价值的文章跟高水平的球友出现，跟我走的时候差不多熊样。 我第一次建仓国海是201704，6.48，到最低点2.75（配股除权前）跌了58%，中间不断的跌了加/加了跌，201810在跌停板两块九上买了最...', 'type': '2', 'source_link': 'http://xueqiu.com/about/mobile-xueqiu', 'edited_at': None, 'pic': '', 'user': {'subscribeable': False, 'remark': None, 'status': 0, 'common_count': 0, 'domain': None, 'description': '看大牛熊，低估买入靠谱标的，等风来。', 'type': '1', 'location': None, 'id': 2269671151, 'url': None, 'screen_name': '北溟有鱼天海风', 'following': False, 'profile': '/2269671151', 'verified': False, 'blocking': False, 'donate_count': 0, 'friends_count': 427, 'followers_count': 8022, 'st_color': '1', 'follow_me': False, 'verified_description': '', 'province': '山东', 'city': '青岛', 'gender': 'm', 'status_count': 353, 'step': 'null', 'blog_description': None, 'verified_type': 0, 'last_status_id': 143048629, 'stats': None, 'recommend': None, 'intro': None, 'stock_status_count': None, 'allow_all_stock': False, 'stocks_count': None, 'name': None, 'group_ids': None, 'name_pinyin': None, 'screenname_pinyin': None, 'block_status': 0, '_source': None, 'verified_realname': False, 'verified_infos': [{'verified_type': '5', 'verified_desc': '用户已完成实名身份认证'}], 'profile_image_url': 'community/20174/1494989068579-1494989068954.jpg,community/20174/1494989068579-1494989068954.jpg!180x180.png,community/20174/1494989068579-1494989068954.jpg!50x50.png,community/20174/1494989068579-1494989068954.jpg!30x30.png', 'photo_domain': '//xavatar.imedao.com/'}, 'retweeted_status': None, 'target': '/2269671151/143048629', 'fragment': None, 'blocked': False, 'blocking': False, 'topic_pic': None, 'topic_symbol': None, 'topic_title': None, 'topic_desc': None, 'donate_count': 0, 'donate_snowcoin': 0, 'view_count': 0, 'offer': None, 'mark': 0, 'comment': None, 'expend': True, 'rqtype': None, 'rqid': 0, 'firstImg': None, 'card': None, 'favorited': False, 'favorited_created_at': None, 'topic_pic_thumbnail': None, 'topic_pic_headOrPad': None, 'topic_pic_thumbnail_small': None, 'canEdit': True, 'shortStatusMultiPic': False, 'timeBefore': '42分钟前', 'mark_desc': None, 'reward_count': 0, 'reward_user_count': 0, 'paid_mention': None, 'is_answer': False, 'is_refused': False, 'is_reward': True, '_source': 'sc_1:5:01810', 'pic_sizes': [], 'like_count': 1, 'liked': False, 'promotion_pic': None, 'promotion_url': None, 'promotion_id': 0, 'quote_cards': None, 'is_ss_multi_pic': False, 'excellent_comments': None, 'source': 'iPhone', 'is_bonus': False}"
143047461,1330041231,1583420813000,产品性价比高一直是小米的标签，但是这也容易带来争议。争议一：性价比高影响企业利润，甚至有低价误国这种无稽之谈。争议二：随着小米推出价格高于3000元甚至4000元的手机，不少米粉开始指责小米价格不再亲民，业界也评价其丧失性价比竞争力。以上争议有的出自米粉，有的出自投资者，也有出于菊花粉。我说呢，小米的价格战略是小米企业战略的核心部分，只有深刻理解营销理论的人才能体会。出于自身与小米关系而导致的左右摇摆的看法都是非客观的。本文不惜笔墨，追根究底性价比战略，期望一言澄清，以正视听。首先要澄清关于性价比的两个误区：一是认为高性价比就是低价。性价比是性能和价格的比值，怎么可以拿低价去替代呢？性价比战略和低价战略完全是两种不同的战略。有些消费者嘴里说的是追求高性价比，其实就是追求低价。这种认知属于典型的心理保护。二是狭义地理解性价比就是产品技术指标和价格的比值。实际上虽然性价比这个名词出自电子产品，但演变至今，其分子指向实际是产品综合性能对消费者产生的使用价值。这里面，还存在不同消费者之间的差异，甚至其分母，也就是价格，对不同消费者的弹性也是不一样的。下面正式拆解性价比。性价比的分子是产品的性能带给消费者使用价值，企业无疑都是要提升这个价值的，其努力方向有两个，一是选择生产更高端的产品，但是这带来了成本的增加，二是判断哪些性能的增加对提升消费者使用价值权重更大。性价比的分母是价格，追求高性价比的企业无疑都要降低这个数值，但是价格的降低造成两个方向的影响，要么影响产品的成本，制约性能，要么降低企业利润。假设忽略不同性能的增加对提升消费者使用价值权提升的权重差异，再假设各企业的利润诉求一致，那么性价比就是性能除以价格。实际上，如今手机为代表的电子产品同质化极高，硬件指标和综合成本几乎成正比，在以上两个假设下，所有企业的所有手机的性价比应该完全一致。至于实际上出现了那么多让我们感觉性价比差异如此之大的手机，是各家企业战略差异导致。如上图所示，产品的性能和价格基本是成正比的，理论上应该是中间那条虚线。但是实际上，是那条曲线。产品的性能最低的时候，会面临两个问题，一是性能不达到一定门槛，产生不了实际价值。比如现在1G内存的手机，虽然还有一定的成本，但是根本没法使用。二是最低端的产品区，很可能市场极小，或者生产企业规模极小，导致固定成本太高。所以每单位成本提升时，刚开始的时候，性能提升是很慢的。性能最高的时候，也会面临问题，一是高端产品往往都是新研发，存在研发的固定成本，另外市场也不大。所以高端区域，每单位成本提升时，性能提升也变慢了。只有在中间区域，性能随着每单成本的提升，增长最显著。这就决定了，选择中端档次的企业，性价比条件是最好的。不同的企业根据行业特性和自身核心竞争力，依据企业战略在上图中选择适合自己的区域。产业链中上游的企业，往往选择高性价比区域，因为下游采购企业一般都是理性的。凡是参与过招标的人都清楚，设定一个性能门槛，或者低价者得，或者评标打分，中标的往往都是中端供应商。这也是我不买产业链中上游企业股票的原因，因为它们往往被迫处于高性价比区，没有实施自身价格战略的自由，也就既无法获得溢价，可能也无法获得无止境的市场份额。而消费品生产企业可以自由地选择价格战略。有的选择低价战略，抢夺最低端的用户市场，典型如春秋航空。有的走高端路线，甚至产品售价远远高出成本和应得利润，实施撇脂策略。走高端路线的企业有的依靠品牌，比如茅台。有的依靠技术能力，比如苹果。走中端或者低端路线的企业，也有可能形成规模优势，巩固自己的价格战略，依靠的是企业的管理和营销能力。以上都是好的战略，这些企业的股票都值得我们购买。但小米是一个特殊的。毫无疑问，小米一直在执行高性价比战略，最初创业到如今，手机、电视以及各类产品都在中端档次。对其当下手机价格突破4千有质疑的人，你要知道近几年手机价位一直在抬升，4千已经不是高端手机了。另外，小米对具体配件的选择也在尽量符合用户使用需要，而精打细算地花着成本，本文就不展开了。小米的轻资产代工模式、减少研发费用的模式都是服务于其性价比战略的。关于这个问题，我几天前引用的《中外管理》中相关文章详尽阐述了，低价误国论可以休矣。然而，做完这些，小米还嫌不够。他还要通过压缩自己利润的方式来降低售价，实施上图中的低毛利策略。我作为一个消费者，已经明显感受到它最后的这个努力带来了产品的不可抗拒。显然这是把能挣的钱让了出去。如果不这么做，销量低一点，但是总利润一定高。为什么要这么做，全世界的上市公司都不会这么做，所有的投资者都反对这么做。只因为一个人的梦想，一个已经功成名就，还要创业的人。如同很多菊花粉所说，没有无缘无故的让利，总会捞回来的。没错，他要的是让小米成为三星那样规模的全球家用电子产品领导者。他要成为乔布斯或者刚刚逝去的前通用CEO杰克·韦尔奇那样的人物。他就像那日本百年老店中绝不轻言涨价的传世工匠。他通过个人权威在一个上市公司内坚持另类的主张，他通过过去10年的飞速发展说服他的同僚和投资人支持他的战略。我作为一个小股东根本无所谓小米通过兑现利润来推高股价，我期望它的市场越来越广阔，早晚它会给我更多的回报。,"{'id': 143047461, 'user_id': 1330041231, 'title': '我为什么认可小米（四）：小米的性价比战略', 'text': '<p>产品性价比高一直是小米的标签，但是这也容易带来争议。争议一：性价比高影响企业利润，甚至有低价误国这种无稽之谈。争议二：随着小米推出价格高于3000元甚至4000元的手机，不少米粉开始指责小米价格不再亲民，业界也评价其丧失性价比竞争力。以上争议有的出自米粉，有的出自投资者，也有出于菊花粉。我说呢，小米的价格战略是小米企业战略的核心部分，只有深刻理解营销理论的人才能体会。出于自身与小米关系而导致的左右摇摆的看法都是非客观的。本文不惜笔墨，追根究底性价比战略，期望一言澄清，以正视听。</p><p>首先要澄清关于性价比的两个误区：一是认为高性价比就是低价。性价比是性能和价格的比值，怎么可以拿低价去替代呢？性价比战略和低价战略完全是两种不同的战略。有些消费者嘴里说的是追求高性价比，其实就是追求低价。这种认知属于典型的心理保护。二是狭义地理解性价比就是产品技术指标和价格的比值。实际上虽然性价比这个名词出自电子产品，但演变至今，其分子指向实际是产品综合性能对消费者产生的使用价值。这里面，还存在不同消费者之间的差异，甚至其分母，也就是价格，对不同消费者的弹性也是不一样的。</p><p>下面正式拆解性价比。性价比的分子是产品的性能带给消费者使用价值，企业无疑都是要提升这个价值的，其努力方向有两个，一是选择生产更高端的产品，但是这带来了成本的增加，二是判断哪些性能的增加对提升消费者使用价值权重更大。性价比的分母是价格，追求高性价比的企业无疑都要降低这个数值，但是价格的降低造成两个方向的影响，要么影响产品的成本，制约性能，要么降低企业利润。假设忽略不同性能的增加对提升消费者使用价值权提升的权重差异，再假设各企业的利润诉求一致，那么性价比就是性能除以价格。实际上，如今手机为代表的电子产品同质化极高，硬件指标和综合成本几乎成正比，在以上两个假设下，所有企业的所有手机的性价比应该完全一致。至于实际上出现了那么多让我们感觉性价比差异如此之大的手机，是各家企业战略差异导致。</p><p><img class=""ke_img"" src=""https://xqimg.imedao.com/170ab3ab1d21165b3fa4f958.png!custom.jpg"" /></p><p>如上图所示，产品的性能和价格基本是成正比的，理论上应该是中间那条虚线。但是实际上，是那条曲线。</p><p>产品的性能最低的时候，会面临两个问题，一是性能不达到一定门槛，产生不了实际价值。比如现在1G内存的手机，虽然还有一定的成本，但是根本没法使用。二是最低端的产品区，很可能市场极小，或者生产企业规模极小，导致固定成本太高。所以每单位成本提升时，刚开始的时候，性能提升是很慢的。性能最高的时候，也会面临问题，一是高端产品往往都是新研发，存在研发的固定成本，另外市场也不大。所以高端区域，每单位成本提升时，性能提升也变慢了。只有在中间区域，性能随着每单成本的提升，增长最显著。这就决定了，选择中端档次的企业，性价比条件是最好的。</p><p>不同的企业根据行业特性和自身核心竞争力，依据企业战略在上图中选择适合自己的区域。产业链中上游的企业，往往选择高性价比区域，因为下游采购企业一般都是理性的。凡是参与过招标的人都清楚，设定一个性能门槛，或者低价者得，或者评标打分，中标的往往都是中端供应商。这也是我不买产业链中上游企业股票的原因，因为它们往往被迫处于高性价比区，没有实施自身价格战略的自由，也就既无法获得溢价，可能也无法获得无止境的市场份额。</p><p>而消费品生产企业可以自由地选择价格战略。有的选择低价战略，抢夺最低端的用户市场，典型如春秋航空。有的走高端路线，甚至产品售价远远高出成本和应得利润，实施撇脂策略。走高端路线的企业有的依靠品牌，比如茅台。有的依靠技术能力，比如苹果。走中端或者低端路线的企业，也有可能形成规模优势，巩固自己的价格战略，依靠的是企业的管理和营销能力。以上都是好的战略，这些企业的股票都值得我们购买。</p><p>但小米是一个特殊的。毫无疑问，小米一直在执行高性价比战略，最初创业到如今，手机、电视以及各类产品都在中端档次。对其当下手机价格突破4千有质疑的人，你要知道近几年手机价位一直在抬升，4千已经不是高端手机了。另外，小米对具体配件的选择也在尽量符合用户使用需要，而精打细算地花着成本，本文就不展开了。小米的轻资产代工模式、减少研发费用的模式都是服务于其性价比战略的。关于这个问题，我几天前引用的《中外管理》中相关文章详尽阐述了，低价误国论可以休矣。</p><p>然而，做完这些，小米还嫌不够。他还要通过压缩自己利润的方式来降低售价，实施上图中的低毛利策略。我作为一个消费者，已经明显感受到它最后的这个努力带来了产品的不可抗拒。显然这是把能挣的钱让了出去。如果不这么做，销量低一点，但是总利润一定高。为什么要这么做，全世界的上市公司都不会这么做，所有的投资者都反对这么做。只因为一个人的梦想，一个已经功成名就，还要创业的人。如同很多菊花粉所说，没有无缘无故的让利，总会捞回来的。没错，他要的是让小米成为三星那样规模的全球家用电子产品领导者。他要成为乔布斯或者刚刚逝去的前通用CEO杰克·韦尔奇那样的人物。他就像那日本百年老店中绝不轻言涨价的传世工匠。他通过个人权威在一个上市公司内坚持另类的主张，他通过过去10年的飞速发展说服他的同僚和投资人支持他的战略。我作为一个小股东根本无所谓小米通过兑现利润来推高股价，我期望它的市场越来越广阔，早晚它会给我更多的回报。</p><p><a href=""http://xueqiu.com/S/01810"" target=""_blank"">$小米集团-W(<span class=\'highlight\'>01810</span>)$</a> <a href=""http://xueqiu.com/S/SH601021"" target=""_blank"">$春秋航空(SH601021)$</a> <a href=""http://xueqiu.com/S/SH600519"" target=""_blank"">$贵州茅台(SH600519)$</a> </p>', 'created_at': 1583420813000, 'retweet_count': 0, 'reply_count': 3, 'fav_count': 0, 'truncated': False, 'commentId': 0, 'retweet_status_id': 0, 'symbol_id': None, 'description': '...投资人支持他的战略。我作为一个小股东根本无所谓小米通过兑现利润来推高股价，我期望它的市场越来越广阔，早晚它会给我更多的回报。 $小米集团-W( <span class=\'highlight\'>01810</span> )$ <a href=""http://xueqiu.com/S/SH601021"" target=""_blank"">$春秋航空(SH601021)$</a> <a href=""http://xueqiu.com/S/SH600519"" target=""_blank"">$贵州茅台(SH600519)$</a> ', 'type': '3', 'source_link': None, 'edited_at': 1583421194000, 'pic': 'https://xqimg.imedao.com/170ab3ab1d21165b3fa4f958.png!thumb.jpg', 'user': {'subscribeable': False, 'remark': None, 'status': 0, 'common_count': 0, 'domain': None, 'description': '我们对企业一无所知，但了解人性。', 'type': '1', 'location': None, 'id': 1330041231, 'url': None, 'screen_name': '南墙幽瓦', 'following': False, 'profile': '/1330041231', 'verified': False, 'blocking': False, 'donate_count': 0, 'friends_count': 15, 'followers_count': 856, 'st_color': '1', 'follow_me': False, 'verified_description': None, 'province': '江苏', 'city': '苏州', 'gender': 'm', 'status_count': 41, 'step': 'null', 'blog_description': None, 'verified_type': 0, 'last_status_id': 143047461, 'stats': None, 'recommend': None, 'intro': None, 'stock_status_count': None, 'allow_all_stock': False, 'stocks_count': None, 'name': None, 'group_ids': None, 'name_pinyin': None, 'screenname_pinyin': None, 'block_status': 0, '_source': None, 'verified_realname': False, 'verified_infos': None, 'profile_image_url': 'community/201811/1546055740032-1546055747533.jpg,community/201811/1546055740032-1546055747533.jpg!180x180.png,community/201811/1546055740032-1546055747533.jpg!50x50.png,community/201811/1546055740032-1546055747533.jpg!30x30.png', 'photo_domain': '//xavatar.imedao.com/'}, 'retweeted_status': None, 'target': '/1330041231/143047461', 'fragment': None, 'blocked': False, 'blocking': False, 'topic_pic': None, 'topic_symbol': None, 'topic_title': None, 'topic_desc': None, 'donate_count': 0, 'donate_snowcoin': 0, 'view_count': 0, 'offer': None, 'mark': 0, 'comment': None, 'expend': True, 'rqtype': None, 'rqid': 0, 'firstImg': 'https://xqimg.imedao.com/170ab3ab1d21165b3fa4f958.png!thumb.jpg', 'card': None, 'favorited': False, 'favorited_created_at': None, 'topic_pic_thumbnail': None, 'topic_pic_headOrPad': None, 'topic_pic_thumbnail_small': None, 'canEdit': True, 'shortStatusMultiPic': False, 'timeBefore': '54分钟前', 'mark_desc': None, 'reward_count': 0, 'reward_user_count': 0, 'paid_mention': None, 'is_answer': False, 'is_refused': False, 'is_reward': True, '_source': 'sc_1:6:01810', 'pic_sizes': [{'width': 858, 'height': 581, 'ratio': '1.48', 'score': 1, 'setScore': True, 'setRatio': True, 'setWidth': True, 'setHeight': True}], 'like_count': 2, 'liked': False, 'promotion_pic': None, 'promotion_url': None, 'promotion_id': 0, 'quote_cards': None, 'is_ss_multi_pic': False, 'excellent_comments': None, 'source': '雪球', 'is_bonus': False}"
143047430,3818775985,1583420795000,回复@远行者: 是的！这篇演讲雷军把小米讲透了。//@远行者:回复@等待就是忍耐:雷军对小米很好的诠释，解开了很多疑点。,"{'id': 143047430, 'user_id': 3818775985, 'title': '', 'text': '回复<a href=""http://xueqiu.com/n/远行者"" target=""_blank"">@远行者</a>: 是的！这篇演讲雷军把小米讲透了。//<a href=""http://xueqiu.com/n/远行者"" target=""_blank"">@远行者</a>:回复<a href=""http://xueqiu.com/n/等待就是忍耐"" target=""_blank"">@等待就是忍耐</a>:雷军对小米很好的诠释，解开了很多疑点。 <a href=""http://xueqiu.com/S/01810"" target=""_blank"">$小米集团-W(<span class=\'highlight\'>01810</span>)$</a>', 'created_at': 1583420795000, 'retweet_count': 0, 'reply_count': 0, 'fav_count': 0, 'truncated': False, 'commentId': 139502015, 'retweet_status_id': 143037703, 'symbol_id': None, 'description': '...远行者 : 是的！这篇演讲雷军把小米讲透了。// <a href=""http://xueqiu.com/n/远行者"" target=""_blank"">@远行者</a> :回复 <a href=""http://xueqiu.com/n/等待就是忍耐"" target=""_blank"">@等待就是忍耐</a> :雷军对小米很好的诠释，解开了很多疑点。 $小米集团-W( <span class=\'highlight\'>01810</span> )$ ', 'type': None, 'source_link': 'http://xueqiu.com/about/mobile-xueqiu', 'edited_at': None, 'pic': '', 'user': {'subscribeable': False, 'remark': None, 'status': 1, 'common_count': 0, 'domain': None, 'description': '马明哲：我们积蓄了这么多年的能量，无论公司还是股票，现在的爆发只是刚刚开始呢，好戏在后面。（2016年股东大会）', 'type': '1', 'location': None, 'id': 3818775985, 'url': None, 'screen_name': '等待就是忍耐', 'following': False, 'profile': '/3818775985', 'verified': False, 'blocking': False, 'donate_count': 0, 'friends_count': 84, 'followers_count': 265, 'st_color': '1', 'follow_me': False, 'verified_description': None, 'province': '浙江', 'city': '杭州', 'gender': 'm', 'status_count': 934, 'step': 'three', 'blog_description': None, 'verified_type': 0, 'last_status_id': 143050445, 'stats': None, 'recommend': None, 'intro': None, 'stock_status_count': None, 'allow_all_stock': False, 'stocks_count': None, 'name': None, 'group_ids': None, 'name_pinyin': None, 'screenname_pinyin': None, 'block_status': 0, '_source': None, 'verified_realname': False, 'verified_infos': None, 'profile_image_url': 'community/20177/1501661898351-1501661898978.jpg,community/20177/1501661898351-1501661898978.jpg!180x180.png,community/20177/1501661898351-1501661898978.jpg!50x50.png,community/20177/1501661898351-1501661898978.jpg!30x30.png', 'photo_domain': '//xavatar.imedao.com/'}, 'retweeted_status': {'id': 143037703, 'user_id': 3818775985, 'title': '', 'text': '<a href=""http://xueqiu.com/S/01810"" target=""_blank"">$小米集团-W(01810)$</a> 雷军清华演讲：小米9年奇迹背后 未来十年增长十倍<br/><br/>2019-04-05 <br/>       4月4日，应清华大学国家金融研究院院长、IMF前副总裁朱民之邀，雷军在清华大学进行了一次演讲，主题是“小米9年：创新、变革与未来”，主要介绍了小米的创立过程和发展模式，并与与会的清华师生、校友与社会各界人士探讨了未来5G、物联网和人工智能的发展。<br/>雷军表示，小米的创立的初心就是就是希望改变中国制造业，做感动人心的好产品，造福世界上每一个人。<br/>“有些大品牌高高在上，他们是上帝，我们不是，我们是真诚对待用户，把他们当朋友。这样的结果就是我们是全球罕见拥有海量粉丝支持的公司，”雷军说。<br/>雷军还透露，他在内部提了一个目标“未来十年，增长十倍”，手机+AIoT就是增长计划背后很重要的逻辑。<br/>以下为雷军清华大学发言实录：<br/>很荣幸参与朱民行长主持的系列讲座，我是第一个被邀请来讲课的中国人。朱民行长给了我一个题目叫“小米9年：创新、变革与未来”。等大家听我讲完，你会发现小米是和你以往的认知是不一样的。<br/>01、小米的初心<br/>九年前在中关村保福寺桥边上银谷大厦一间很小办公室，10来个人7、8条枪，一起喝了碗小米粥，开始“闹革命”。之前我们这些人从来没有做过制造业，也从来没有做过硬件，我们主要来自于金山、谷歌、微软一群做互联网的人，我们看到了一个很大的问题解决不了，什么问题呢？就是咱们中国是世界制造工厂，但是没有什么真正的好产品，满大街都是山寨手机，这就是在我们中国发生的现象。<br/>说实话，40岁创业和20岁创业最大的差别是，一上来不认为我们自己会成功。我认为成功概率很低，如果我们全力去试，输了也没有什么，输了至少自己不遗憾。初期创业一年半我是隐姓埋名，因为我担心我一干这个公司，满世界都关注，给我自己压力很大，我跟大家说偷偷地干，这样自己压力比较小，干砸了也没有什么。<br/>为什么叫小米呢？原先我们想叫大米，后来要去高大上，变成小米了。叫小米以后，我们决定继承小米加步枪的精神开始艰苦创业，我们在初期的时候，像中关村几万家普通的创业公司一样，租了一个很小的办公室，规定出差只能住如家酒店，飞机只能坐经济舱。刚开始我投了一千万人民币，一千万人民币注册对于很多创业者来说是很大的数字，但是各位如果做手机，一千万人民币真的是少得可怜的一点钱。<br/>小米的几位创始人在各自的领域里面都是挺牛的一群人，凝聚在一起是为了一个与众不同的梦想，就是希望用我们的能力改变中国制造业，做感动人心的好产品，造福世界上每一个人。<br/>这就是我们的初心。<br/>9年前我们进入的第一个行业是智能手机，这个行业当时有苹果、三星、中华酷联，全是巨无霸，一个中关村10来个人的小公司，杀到一个全世界竞争最激烈的战场，奇迹发生了。小米只用了两年半，做到了中国第一、世界第三。<br/>然而我们保持了两年中国第一的位置以后接着下滑了，其实原因很简单，光有互联网思想远远不够，如果我们对传统制造行业的经验缺少敬畏之心，缺少愿意学习的心态，我们打不赢这场仗。无论我们掌握了多么先进的武器，如果没有良好的基础一定会输的。<br/>从2016年开始，小米强调补课，向传统产业学习，放下身段逐个对标。结果第二个奇迹发生了，历史上在手机行业没有任何一家大公司销量一旦下滑还能够活得过来，一旦失去信心以后会加速下滑，基本上是死亡深渊。小米从2016年滑到谷底，2017年售出了9400万台，2018年售出了1.18亿台。在2017年的时候，我们重回了世界第四，去年突破了1亿台。<br/>我们前不久公布了去年财报，创办第八年的收入1749亿，同比增长52.6%，利润86亿，同比增长59.5%，这个业绩是在去年中美贸易摩擦下行压力进一步加剧的前提下创造的，去年对制造业的压力是巨大无比，但是在如此大的压力下，小米的成绩单都是呈现50%、60%的增长，而且小米一个8年多的公司，国际收入占比超过了40%，进入了80多个国家和市场，在30个国家的智能手机市场排到前五位。<br/>比如说像印度，进去三年以后也排到印度第一位，现在连续八个季度印度第一，我们在印度占了30%的份额。在全球大市场里第一，中国第一大市场，美国第二大，印度第三大，美国先放一放，我们把印度市场做起来以后，最近主攻方向是欧洲，在过去一年里面，我们在欧洲销量涨了10倍，在西班牙等一些国家排到前三位了，这说明小米模式被充分验证了。<br/>02、小米奇迹的背后：创新驱动<br/>小米的创新点在什么地方呢？其实李克强总理说了一个词叫“互联网+”，我们就是最典型的互联网+，用互联网模式做硬件，是一种高维度的创新。我们是怎么做的呢？因为今天时间有限，我就只讲几个最关键的点：<br/>第一，和用户真心交朋友。这个话很简单，但是不容易做，因为在中国有句古话叫无商不奸，所以你说你真心和用户交朋友，在中国很难做到的。但是互联网有一个最大的好处，我们在初期的时候我自己带头所有人都在社区里倾听用户声音，收集了2.1亿个帖子的建议和资料，我们把互联网快速迭代引入了硬件产品的研发，把手机系统做到了每周迭代，当时做出来的时候很多人很惊讶，我们坚持到今天主要是针对两、三百万的铁杆粉丝做每周升级。<br/>我亲自带头发微博，在今年小米9发布前的一个星期，我统计了一下我发了97条微博。于是米粉亲切评价说我是小米公司第一数码博主。这些全是我自己发的。其实我每天很忙很忙，用所有的碎片时间跟用户沟通，还每条用户的反馈都看了，这是一个巨大的工作量，但是我认为这是我在小米最主要的工作。今天上午我们就用户意见改善的会议开了三个小时，100多个高级主管在一起，我说我们现在做的事情就是坚持初心，坚持跟用户交朋友，坚持根据用户的意见去改进。不要觉得我们大了，不要觉得我们怎么样了，这些都不重要，重要的是我们做的是不是用户要的东西。<br/>有些大品牌高高在上，他们是上帝，我们不是，我们是真诚对待用户，把他们当朋友。这样的结果就是我们是全球罕见拥有海量粉丝支持的公司。<br/>去年5月在法国的第一家小米之家店开业时下起冰雹，很多人坚持排队。在西班牙更是如此，西班牙人对小米疯狂得不得了，我认为这个模式对全世界每个人都有价值，就像我当初提的，让全球每一个人都能享受科技带来美好生活，每个人都可以感受得到。<br/>这一套模式的核心是“感动人心、价格厚道”，其实每一个公司做起来的时候都有一段价格厚道的时候，每个公司在起来的时候都用过一段价格战，都是这么一步一步成长起来的，但是这些公司把性价比当成战术没有当成价值观。我也特别害怕10年后、20年后小米也变成我曾经憎恨的公司，于是在上市之前我们出了一套“宪法”，小米对用户永久性承诺，小米每年整体硬件业务的综合净利率永远不会超过5%，如果超过我们都将回馈给用户。我们把这一条做成规定，每年请审计公司审计我们的利润率。<br/>我曾经研究过什么叫创新，包括怎么创新。只有两件事叫创新：一是做别人没有做过的事情。二是做别人做过了失败的事情。这样的事情肯定不好做，这样的事情做的结果99%是输，我们国家鼓励创新，我们每个人要创新，但是一定要意识到创新的后果是绝大部分都会输，一个创新氛围应该是一个包容失败，包容错误，包容这些东西的文化才是创新的前提。我们不能成王败寇，成了你就是英雄，输了你就傻，你就笨，那样的话就没有人创新了。<br/>我们和用户交朋友，是要达到一种境界，即让用户愿意闭着眼睛买小米产品。这个要求其实很高。只要需要，我不关心它的设计，不关心它的质量，也不关心体验，更不用关心它的价钱，因为都是最好的。小米的目标是有朝一天这个用户能够闭这眼睛买我的东西，做到了我就赢了。一上来我就保证我就赚了1、2个点而已，这是一个很极端的做法，但是一个公司有几万人的时候，你怎么保证所有人想的一样呢？<br/>我在内部一直讲“优秀的公司获得利润，伟大的公司赢得人心”。我认为小米走的是伟大之路，万一我们成了，有可能成为这个星球上最伟大的公司。我们不要斤斤计较利润的高和低，我们要关心的是用户支持不支持我们，喜欢不喜欢我们，是不是喜欢我们的变多了，这是我们关心的事情。<br/>第二，做感动人心的好产品。<br/>想和用户交朋友不仅仅是态度，不仅仅是利润率，更重要是做出感动人心的好产品。初期我们整个创始人班子100%全部是研发工程师背景，平均20年行业经验。在创立这个公司的时候，所有人都是技术背景，整天在想怎么做好产品。做好产品最核心的是创新和研发，小米的创新力被世界各个知名机构和媒体评价，在中国的创新都是顶级的，前不久财富杂志评选小米在中国的创新排第三名。<br/>再说一个关键问题，我们才干了8年，我们去年在研发费用上花了58亿元人民币，同比增长83%。我举几个例子说明小米在技术创新方面的成绩：<br/>1、人工智能。人工智能音箱是今天整个行业的制高点，小爱同学出货量超过1000万台，在中国排在数一数二的地位，小爱同学语音助理月活跃3880万排在中国第一，这种智能交互人工智能系统用的人越多，系统越好。今天随着我们上千名人工智能方面的工程师不断改进，用的人越来越多，我们人工智能产品的实用性就越来越好。<br/>日经中文网统计了一下，小米在AI专利申请量全世界排在11位，这个数字表明了小米在人工智能领域的投入远超大家想象。<br/>今天几乎大部分人用的是全面屏手机，这个概念是由小米发明的，小米拥有这个领域100多项专利，我们在这个方面做了大量的预研工作，带动了全行业的换代和升级。<br/>5、6年前小米开始做手机芯片，手机芯片需要很长时间，两年前小米发布了第一代芯片。<br/>在5G研发方面，今年2月份小米在巴塞罗那世界移动通信大会 （简称：MWC）上发布了第一部5G手机，所以小米在研发上下了很多功夫投入很大。<br/>2、设计。小米MIX手机这一款现代消费电子产品，先后被芬兰国家艺术馆、法国蓬皮杜艺术中心和德国慕尼黑国际设计博物馆收藏，这表达了小米在产品上设计水平的能力。我们横扫了这个星球上所有最高级别的设计奖，包括IDEA设计金奖、红点最佳设计奖、IF设计金奖和日本优良设计金奖。搞不好我们是中国公司里面拿设计奖最多的公司，总计拿了200多项。。<br/>3、品质。我在内部讲的一句话，后来成了金句：“创新决定我们能飞得有多高，品质决定我们能够走多远。”我自己亲自参加公司内部很多次质量大会，我在那些会上引用总理讲的话“质量之魂存于匠心”。过去一年多，凡是中国有质量奖评选的活动我们几乎都参与了，而且基本上我们都获奖了。<br/>第三，高效率的商业模型。<br/>中国的产品为什么做不好还有最重要一点是效率。各个环节消费很低，我们做一个好产品以后成本很高，所以需要卖的非常贵才能挣钱。提高效率以后，我可以卖的非常便宜，但是把钱主要用在研发上，这样可以很便宜价钱买到其实质量非常好的产品。谈到效率这个问题，我再科普一下商品价格构成。<br/>一个商品的价格包括原材料和制造成本、研发设计分摊成本、市场推广及广告、销售及渠道成本和利润。你们买到任何一个商品最核心是两块：原材料及制造研发、设计成本。我们的关键点是怎么来降低和消解这五项核心成本。我们要把这个产品做好，原材料及制造研发、设计成本这两项省不掉的，甚至你可能要做的更夸张。我在创办小米的时候，10来个人，一千万人民币的公司，初期选择供应商，不是苹果供应商不用，不是三星旗舰手机供应商不用，门槛巨高无比。为什么小米手机在初期一出厂的时候就是世界顶级产品和品质？初期小米出厂的时候，平均售价是1999元人民币，当时旗舰手机都卖4千到7千的时候，我们2千块钱手机的成本几乎都用在生产和制造上。<br/>研发分摊成本是什么呢？它不是研发成本总量，是研发分摊到每一个东西上的成本，是研发做一款产品的成本。我刚开始做手机的时候，同行每年做100款手机，每一款手机花1000万人民币，花了10亿人民币研发，你们会觉得它很厉害，实际上我花了10倍，我只做一款手机，花10倍就是1亿人民币，我总的研发成本就是它的1/10，但是我花在一款产品上的钱是它的10倍。结果我这款手机拿上来，小米第一代卖了760万台。我们做很少的SKU，投的研发的钱却是很多的。<br/>我办小米的时候又做了一个大胆的尝试，前五年零广告费，一分钱没有花过。关于销售和渠道成本，我又做了一个很夸张的动作，就是前店后厂。左边生产了一部手机，右边我开一个网站叫小米网直接卖了。<br/>接着问题是什么呢？我们只挣一两个点，我们怎么能够找到这个星球上最优秀的人，我们怎么能够有足够研发投资，甚至有足够外部投资帮助我们渡过难关。这个事情很难吧，因为利润很少，你很容易经营不好就破产了。我提出来“铁人三项”，实际上我们既做硬件，也做软件，也做互联网，也做服务，我们得垂直整合，在你看不到，在你没有意识到的地方我们也要认真经营，所以我们采用了互联网模式来做实体经济。<br/>我研究过Costco的利润，这么低的毛利基本上不可能挣钱，最高毛利不超过14%，综合毛利在7%到8%，它居然不亏钱，我就琢磨怎么挣钱，第一个挣钱，它开始收会员费，后来开始推广Costco信用卡，因为美国信用卡扣2到3个点，所以它实际上是早期互联网+的模型。<br/>我们去年互联网服务收入到了160亿，同比增长61.2%，这基本上就是我们的利润来源，前年是99亿，去年160亿，互联网服务的增长还是非常快。<br/>总结下我的观点：性价比不是战术，性价比是一种价值观，性价比是一种商业模式，性价比是一种核心能力，背后是对效率的极致追求和死磕自己的决心。<br/>这里面含有两条重要的内涵：第一条，极致的效率；第二条，严于律己死磕自己。<br/>03、小米的未来：手机+AIoT<br/>对于小米，未来就是5G的春天，因为所有人的手机都要换5G。我更看好是5G+AI+IoT，我认为这是下一代超级互联网，能够把所有的东西连起来。小米五年多前开始做物联网，现在做到世界第一了，但是这个领域会继续加速。去年年底小米定了未来十年战略是手机+AIoT双引擎，我们承诺在未来五年里面投资100个亿在AIoT。<br/>我在过去9年在管理上也做了很多大胆的尝试。我们叫去title（职位）；去KPI（关键业绩指标）；去管理。我有这些做法是因为我管公司20年以后想试一下，有没有真正高效的管理。我发现管理是一个悖论，管理就是控制风险，意味着不出现最坏的结果不翻船，但是也不保证出现最好的结果。<br/>十年前我就在想，有没有比大家传统意义上理解的管理更好的管理？后来我想明白了，更好的管理就是不管理。不管理比管理难很多，核心是找一帮优秀的人，找一帮不需要管理的人，塑造共同的愿景和目标，给足够激励就行了。什么样的人不需要管理？要有强大责任心、强大的自我驱动力，强大的梦想和抱负，心甘情愿每天12点下班，不用老板看着。我们小米刚开始就是这样的，我们找了一群非常热忱的人来做，这个模式干了八年以后，取得了巨大成功之后，我们还是需要很冷静地看待，它的问题是你怎么保证2万3千人都这样，他们散布在全球80多个国家，所以我们怎么能够继承发扬小米创业时候的优良特点，结合今天传统通用流程管理，怎么相结合，同时要小心过度管理。<br/>所以，我们又在摸索小米在第二个阶段创业的时候管理模式，干了很多事情：<br/>1、聚焦技术创新。<br/>2、强化总部管理职能，设立组织部和参谋部。<br/>3、梳理业务，提拔年轻干部。<br/>我在内部讲过，其实流程和规范有一点像练剑，手中有剑，我们怎么练到手中无剑心中有剑，怎么升华到无剑无我的阶段，这是武功最高的境界。在初期的时候，我们公司也很少有会议，基本上不怎么开会，这样干了很多年，现在在一步一步导入管理，因为实验完了以后，管一两千人没有问题，到几万人的话，还是需要怎么结合传统管理经验，我们现在正在探索的过程当中。<br/>我在内部提了一个目标“未来十年，增长十倍”。手机+AIoT就是增长计划背后很重要的逻辑。', 'created_at': 1583415919000, 'retweet_count': 12, 'reply_count': 11, 'fav_count': 14, 'truncated': False, 'commentId': 0, 'retweet_status_id': 0, 'symbol_id': None, 'description': '<a href=""http://xueqiu.com/S/01810"" target=""_blank"">$小米集团-W(01810)$</a> 雷军清华演讲：小米9年奇迹背后 未来十年增长十倍2019-04-05 4月4日，应清华大学国家金融研究院院长、IMF前副总裁朱民之邀，雷军在清华大学进行了一次演讲，主题是“小米9年：创新、变革与未来”，主要介绍了小米的创立过程和发展模式，并与与会的清华师生、校友与社会各界人...', 'type': '2', 'source_link': 'http://xueqiu.com/about/mobile-xueqiu', 'edited_at': None, 'pic': '', 'user': {'subscribeable': False, 'remark': None, 'status': 1, 'common_count': 0, 'domain': None, 'description': '马明哲：我们积蓄了这么多年的能量，无论公司还是股票，现在的爆发只是刚刚开始呢，好戏在后面。（2016年股东大会）', 'type': '1', 'location': None, 'id': 3818775985, 'url': None, 'screen_name': '等待就是忍耐', 'following': False, 'profile': '/3818775985', 'verified': False, 'blocking': False, 'donate_count': 0, 'friends_count': 84, 'followers_count': 265, 'st_color': '1', 'follow_me': False, 'verified_description': None, 'province': '浙江', 'city': '杭州', 'gender': 'm', 'status_count': 934, 'step': 'three', 'blog_description': None, 'verified_type': 0, 'last_status_id': 143050445, 'stats': None, 'recommend': None, 'intro': None, 'stock_status_count': None, 'allow_all_stock': False, 'stocks_count': None, 'name': None, 'group_ids': None, 'name_pinyin': None, 'screenname_pinyin': None, 'block_status': 0, '_source': None, 'verified_realname': False, 'verified_infos': None, 'profile_image_url': 'community/20177/1501661898351-1501661898978.jpg,community/20177/1501661898351-1501661898978.jpg!180x180.png,community/20177/1501661898351-1501661898978.jpg!50x50.png,community/20177/1501661898351-1501661898978.jpg!30x30.png', 'photo_domain': '//xavatar.imedao.com/'}, 'retweeted_status': None, 'target': '/3818775985/143037703', 'fragment': None, 'blocked': False, 'blocking': False, 'topic_pic': None, 'topic_symbol': None, 'topic_title': None, 'topic_desc': None, 'donate_count': 0, 'donate_snowcoin': 0, 'view_count': 7919, 'offer': None, 'mark': 0, 'comment': None, 'expend': True, 'rqtype': None, 'rqid': 0, 'firstImg': None, 'card': None, 'favorited': False, 'favorited_created_at': None, 'topic_pic_thumbnail': None, 'topic_pic_headOrPad': None, 'topic_pic_thumbnail_small': None, 'canEdit': True, 'shortStatusMultiPic': False, 'timeBefore': '03-05 21:45', 'mark_desc': None, 'reward_count': 0, 'reward_user_count': 0, 'paid_mention': None, 'is_answer': False, 'is_refused': False, 'is_reward': True, '_source': None, 'pic_sizes': [], 'like_count': 12, 'liked': False, 'promotion_pic': None, 'promotion_url': None, 'promotion_id': 0, 'quote_cards': None, 'is_ss_multi_pic': False, 'excellent_comments': None, 'source': 'iPad', 'is_bonus': False}, 'target': '/3818775985/143047430', 'fragment': None, 'blocked': False, 'blocking': False, 'topic_pic': None, 'topic_symbol': None, 'topic_title': None, 'topic_desc': None, 'donate_count': 0, 'donate_snowcoin': 0, 'view_count': 0, 'offer': None, 'mark': 0, 'comment': None, 'expend': False, 'rqtype': None, 'rqid': 0, 'firstImg': None, 'card': None, 'favorited': False, 'favorited_created_at': None, 'topic_pic_thumbnail': None, 'topic_pic_headOrPad': None, 'topic_pic_thumbnail_small': None, 'canEdit': False, 'shortStatusMultiPic': False, 'timeBefore': '54分钟前', 'mark_desc': None, 'reward_count': 0, 'reward_user_count': 0, 'paid_mention': None, 'is_answer': False, 'is_refused': False, 'is_reward': True, '_source': 'sc_1:7:01810', 'pic_sizes': [], 'like_count': 0, 'liked': False, 'promotion_pic': None, 'promotion_url': None, 'promotion_id': 0, 'quote_cards': None, 'is_ss_multi_pic': False, 'excellent_comments': None, 'source': 'iPad', 'is_bonus': False}"
143045302,4706271602,1583419795000,顶级商业大佬里，军军算是互动最多、最透明的了吧？为什么还是有很多人讨厌他？&nbsp;,"{'id': 143045302, 'user_id': 4706271602, 'title': '', 'text': '<p>顶级商业大佬里，军军算是互动最多、最透明的了吧？为什么还是有很多人讨厌他？<a href=""http://xueqiu.com/S/01810"" target=""_blank"">$小米集团-W(<span class=\'highlight\'>01810</span>)$</a>&nbsp;</p><img src=""https://xqimg.imedao.com/170ab2b4b56114c13fec7c61.png!custom.jpg"" class=""ke_img"" >', 'created_at': 1583419795000, 'retweet_count': 0, 'reply_count': 6, 'fav_count': 0, 'truncated': False, 'commentId': 0, 'retweet_status_id': 0, 'symbol_id': None, 'description': "" 顶级商业大佬里，军军算是互动最多、最透明的了吧？为什么还是有很多人讨厌他？ $小米集团-W( <span class='highlight'>01810</span> )$   "", 'type': '0', 'source_link': None, 'edited_at': None, 'pic': 'https://xqimg.imedao.com/170ab2b4b56114c13fec7c61.png!thumb.jpg', 'user': {'subscribeable': False, 'remark': None, 'status': 2, 'common_count': 0, 'domain': None, 'description': '如果觉得我说的不对，请用数据和事实来打我脸，而不是想当然和情绪。', 'type': '1', 'location': None, 'id': 4706271602, 'url': None, 'screen_name': '小米情报官', 'following': False, 'profile': '/4706271602', 'verified': False, 'blocking': False, 'donate_count': 0, 'friends_count': 12, 'followers_count': 119, 'st_color': '1', 'follow_me': False, 'verified_description': None, 'province': '北京', 'city': '不限', 'gender': 'm', 'status_count': 174, 'step': 'null', 'blog_description': None, 'verified_type': 0, 'last_status_id': 143045302, 'stats': None, 'recommend': None, 'intro': None, 'stock_status_count': None, 'allow_all_stock': False, 'stocks_count': None, 'name': None, 'group_ids': None, 'name_pinyin': None, 'screenname_pinyin': None, 'block_status': 0, '_source': None, 'verified_realname': False, 'verified_infos': None, 'profile_image_url': 'community/20201/1582703078929-1582703083268.jpg,community/20201/1582703078929-1582703083268.jpg!180x180.png,community/20201/1582703078929-1582703083268.jpg!50x50.png,community/20201/1582703078929-1582703083268.jpg!30x30.png', 'photo_domain': '//xavatar.imedao.com/'}, 'retweeted_status': None, 'target': '/4706271602/143045302', 'fragment': None, 'blocked': False, 'blocking': False, 'topic_pic': None, 'topic_symbol': None, 'topic_title': None, 'topic_desc': None, 'donate_count': 0, 'donate_snowcoin': 0, 'view_count': 0, 'offer': None, 'mark': 0, 'comment': None, 'expend': True, 'rqtype': None, 'rqid': 0, 'firstImg': 'https://xqimg.imedao.com/170ab2b4b56114c13fec7c61.png!thumb.jpg', 'card': None, 'favorited': False, 'favorited_created_at': None, 'topic_pic_thumbnail': None, 'topic_pic_headOrPad': None, 'topic_pic_thumbnail_small': None, 'canEdit': True, 'shortStatusMultiPic': False, 'timeBefore': '03-05 22:49', 'mark_desc': None, 'reward_count': 0, 'reward_user_count': 0, 'paid_mention': None, 'is_answer': False, 'is_refused': False, 'is_reward': True, '_source': 'sc_1:8:01810', 'pic_sizes': [{'width': 690, 'height': 971, 'ratio': '0.71', 'score': 1, 'setScore': True, 'setRatio': True, 'setWidth': True, 'setHeight': True}], 'like_count': 1, 'liked': False, 'promotion_pic': None, 'promotion_url': None, 'promotion_id': 0, 'quote_cards': None, 'is_ss_multi_pic': False, 'excellent_comments': None, 'source': '雪球', 'is_bonus': False}"
143045191,4658260427,1583419735000,雷军对小米很好的诠释，解开了很多疑点。,"{'id': 143045191, 'user_id': 4658260427, 'title': '', 'text': '雷军对小米很好的诠释，解开了很多疑点。 <a href=""http://xueqiu.com/S/01810"" target=""_blank"">$小米集团-W(<span class=\'highlight\'>01810</span>)$</a>', 'created_at': 1583419735000, 'retweet_count': 0, 'reply_count': 1, 'fav_count': 0, 'truncated': False, 'commentId': 139499701, 'retweet_status_id': 143037703, 'symbol_id': None, 'description': ""雷军对小米很好的诠释，解开了很多疑点。 $小米集团-W( <span class='highlight'>01810</span> )$ "", 'type': None, 'source_link': 'http://xueqiu.com/about/mobile-xueqiu', 'edited_at': None, 'pic': '', 'user': {'subscribeable': False, 'remark': None, 'status': 1, 'common_count': 0, 'domain': None, 'description': '安全边际下的成长价值投资，顺势而为，做时间的朋友。', 'type': '1', 'location': None, 'id': 4658260427, 'url': None, 'screen_name': '远行者', 'following': False, 'profile': '/4658260427', 'verified': False, 'blocking': False, 'donate_count': 0, 'friends_count': 395, 'followers_count': 4372, 'st_color': '1', 'follow_me': False, 'verified_description': None, 'province': '香港', 'city': '', 'gender': 'm', 'status_count': 4842, 'step': 'three', 'blog_description': None, 'verified_type': 0, 'last_status_id': 143045191, 'stats': None, 'recommend': None, 'intro': None, 'stock_status_count': None, 'allow_all_stock': False, 'stocks_count': None, 'name': None, 'group_ids': None, 'name_pinyin': None, 'screenname_pinyin': None, 'block_status': 0, '_source': None, 'verified_realname': False, 'verified_infos': None, 'profile_image_url': 'community/20146/1405049466936-1405049467101.jpg,community/20146/1405049466936-1405049467101.jpg!180x180.png,community/20146/1405049466936-1405049467101.jpg!50x50.png,community/20146/1405049466936-1405049467101.jpg!30x30.png', 'photo_domain': '//xavatar.imedao.com/'}, 'retweeted_status': {'id': 143037703, 'user_id': 3818775985, 'title': '', 'text': '<a href=""http://xueqiu.com/S/01810"" target=""_blank"">$小米集团-W(01810)$</a> 雷军清华演讲：小米9年奇迹背后 未来十年增长十倍<br/><br/>2019-04-05 <br/>       4月4日，应清华大学国家金融研究院院长、IMF前副总裁朱民之邀，雷军在清华大学进行了一次演讲，主题是“小米9年：创新、变革与未来”，主要介绍了小米的创立过程和发展模式，并与与会的清华师生、校友与社会各界人士探讨了未来5G、物联网和人工智能的发展。<br/>雷军表示，小米的创立的初心就是就是希望改变中国制造业，做感动人心的好产品，造福世界上每一个人。<br/>“有些大品牌高高在上，他们是上帝，我们不是，我们是真诚对待用户，把他们当朋友。这样的结果就是我们是全球罕见拥有海量粉丝支持的公司，”雷军说。<br/>雷军还透露，他在内部提了一个目标“未来十年，增长十倍”，手机+AIoT就是增长计划背后很重要的逻辑。<br/>以下为雷军清华大学发言实录：<br/>很荣幸参与朱民行长主持的系列讲座，我是第一个被邀请来讲课的中国人。朱民行长给了我一个题目叫“小米9年：创新、变革与未来”。等大家听我讲完，你会发现小米是和你以往的认知是不一样的。<br/>01、小米的初心<br/>九年前在中关村保福寺桥边上银谷大厦一间很小办公室，10来个人7、8条枪，一起喝了碗小米粥，开始“闹革命”。之前我们这些人从来没有做过制造业，也从来没有做过硬件，我们主要来自于金山、谷歌、微软一群做互联网的人，我们看到了一个很大的问题解决不了，什么问题呢？就是咱们中国是世界制造工厂，但是没有什么真正的好产品，满大街都是山寨手机，这就是在我们中国发生的现象。<br/>说实话，40岁创业和20岁创业最大的差别是，一上来不认为我们自己会成功。我认为成功概率很低，如果我们全力去试，输了也没有什么，输了至少自己不遗憾。初期创业一年半我是隐姓埋名，因为我担心我一干这个公司，满世界都关注，给我自己压力很大，我跟大家说偷偷地干，这样自己压力比较小，干砸了也没有什么。<br/>为什么叫小米呢？原先我们想叫大米，后来要去高大上，变成小米了。叫小米以后，我们决定继承小米加步枪的精神开始艰苦创业，我们在初期的时候，像中关村几万家普通的创业公司一样，租了一个很小的办公室，规定出差只能住如家酒店，飞机只能坐经济舱。刚开始我投了一千万人民币，一千万人民币注册对于很多创业者来说是很大的数字，但是各位如果做手机，一千万人民币真的是少得可怜的一点钱。<br/>小米的几位创始人在各自的领域里面都是挺牛的一群人，凝聚在一起是为了一个与众不同的梦想，就是希望用我们的能力改变中国制造业，做感动人心的好产品，造福世界上每一个人。<br/>这就是我们的初心。<br/>9年前我们进入的第一个行业是智能手机，这个行业当时有苹果、三星、中华酷联，全是巨无霸，一个中关村10来个人的小公司，杀到一个全世界竞争最激烈的战场，奇迹发生了。小米只用了两年半，做到了中国第一、世界第三。<br/>然而我们保持了两年中国第一的位置以后接着下滑了，其实原因很简单，光有互联网思想远远不够，如果我们对传统制造行业的经验缺少敬畏之心，缺少愿意学习的心态，我们打不赢这场仗。无论我们掌握了多么先进的武器，如果没有良好的基础一定会输的。<br/>从2016年开始，小米强调补课，向传统产业学习，放下身段逐个对标。结果第二个奇迹发生了，历史上在手机行业没有任何一家大公司销量一旦下滑还能够活得过来，一旦失去信心以后会加速下滑，基本上是死亡深渊。小米从2016年滑到谷底，2017年售出了9400万台，2018年售出了1.18亿台。在2017年的时候，我们重回了世界第四，去年突破了1亿台。<br/>我们前不久公布了去年财报，创办第八年的收入1749亿，同比增长52.6%，利润86亿，同比增长59.5%，这个业绩是在去年中美贸易摩擦下行压力进一步加剧的前提下创造的，去年对制造业的压力是巨大无比，但是在如此大的压力下，小米的成绩单都是呈现50%、60%的增长，而且小米一个8年多的公司，国际收入占比超过了40%，进入了80多个国家和市场，在30个国家的智能手机市场排到前五位。<br/>比如说像印度，进去三年以后也排到印度第一位，现在连续八个季度印度第一，我们在印度占了30%的份额。在全球大市场里第一，中国第一大市场，美国第二大，印度第三大，美国先放一放，我们把印度市场做起来以后，最近主攻方向是欧洲，在过去一年里面，我们在欧洲销量涨了10倍，在西班牙等一些国家排到前三位了，这说明小米模式被充分验证了。<br/>02、小米奇迹的背后：创新驱动<br/>小米的创新点在什么地方呢？其实李克强总理说了一个词叫“互联网+”，我们就是最典型的互联网+，用互联网模式做硬件，是一种高维度的创新。我们是怎么做的呢？因为今天时间有限，我就只讲几个最关键的点：<br/>第一，和用户真心交朋友。这个话很简单，但是不容易做，因为在中国有句古话叫无商不奸，所以你说你真心和用户交朋友，在中国很难做到的。但是互联网有一个最大的好处，我们在初期的时候我自己带头所有人都在社区里倾听用户声音，收集了2.1亿个帖子的建议和资料，我们把互联网快速迭代引入了硬件产品的研发，把手机系统做到了每周迭代，当时做出来的时候很多人很惊讶，我们坚持到今天主要是针对两、三百万的铁杆粉丝做每周升级。<br/>我亲自带头发微博，在今年小米9发布前的一个星期，我统计了一下我发了97条微博。于是米粉亲切评价说我是小米公司第一数码博主。这些全是我自己发的。其实我每天很忙很忙，用所有的碎片时间跟用户沟通，还每条用户的反馈都看了，这是一个巨大的工作量，但是我认为这是我在小米最主要的工作。今天上午我们就用户意见改善的会议开了三个小时，100多个高级主管在一起，我说我们现在做的事情就是坚持初心，坚持跟用户交朋友，坚持根据用户的意见去改进。不要觉得我们大了，不要觉得我们怎么样了，这些都不重要，重要的是我们做的是不是用户要的东西。<br/>有些大品牌高高在上，他们是上帝，我们不是，我们是真诚对待用户，把他们当朋友。这样的结果就是我们是全球罕见拥有海量粉丝支持的公司。<br/>去年5月在法国的第一家小米之家店开业时下起冰雹，很多人坚持排队。在西班牙更是如此，西班牙人对小米疯狂得不得了，我认为这个模式对全世界每个人都有价值，就像我当初提的，让全球每一个人都能享受科技带来美好生活，每个人都可以感受得到。<br/>这一套模式的核心是“感动人心、价格厚道”，其实每一个公司做起来的时候都有一段价格厚道的时候，每个公司在起来的时候都用过一段价格战，都是这么一步一步成长起来的，但是这些公司把性价比当成战术没有当成价值观。我也特别害怕10年后、20年后小米也变成我曾经憎恨的公司，于是在上市之前我们出了一套“宪法”，小米对用户永久性承诺，小米每年整体硬件业务的综合净利率永远不会超过5%，如果超过我们都将回馈给用户。我们把这一条做成规定，每年请审计公司审计我们的利润率。<br/>我曾经研究过什么叫创新，包括怎么创新。只有两件事叫创新：一是做别人没有做过的事情。二是做别人做过了失败的事情。这样的事情肯定不好做，这样的事情做的结果99%是输，我们国家鼓励创新，我们每个人要创新，但是一定要意识到创新的后果是绝大部分都会输，一个创新氛围应该是一个包容失败，包容错误，包容这些东西的文化才是创新的前提。我们不能成王败寇，成了你就是英雄，输了你就傻，你就笨，那样的话就没有人创新了。<br/>我们和用户交朋友，是要达到一种境界，即让用户愿意闭着眼睛买小米产品。这个要求其实很高。只要需要，我不关心它的设计，不关心它的质量，也不关心体验，更不用关心它的价钱，因为都是最好的。小米的目标是有朝一天这个用户能够闭这眼睛买我的东西，做到了我就赢了。一上来我就保证我就赚了1、2个点而已，这是一个很极端的做法，但是一个公司有几万人的时候，你怎么保证所有人想的一样呢？<br/>我在内部一直讲“优秀的公司获得利润，伟大的公司赢得人心”。我认为小米走的是伟大之路，万一我们成了，有可能成为这个星球上最伟大的公司。我们不要斤斤计较利润的高和低，我们要关心的是用户支持不支持我们，喜欢不喜欢我们，是不是喜欢我们的变多了，这是我们关心的事情。<br/>第二，做感动人心的好产品。<br/>想和用户交朋友不仅仅是态度，不仅仅是利润率，更重要是做出感动人心的好产品。初期我们整个创始人班子100%全部是研发工程师背景，平均20年行业经验。在创立这个公司的时候，所有人都是技术背景，整天在想怎么做好产品。做好产品最核心的是创新和研发，小米的创新力被世界各个知名机构和媒体评价，在中国的创新都是顶级的，前不久财富杂志评选小米在中国的创新排第三名。<br/>再说一个关键问题，我们才干了8年，我们去年在研发费用上花了58亿元人民币，同比增长83%。我举几个例子说明小米在技术创新方面的成绩：<br/>1、人工智能。人工智能音箱是今天整个行业的制高点，小爱同学出货量超过1000万台，在中国排在数一数二的地位，小爱同学语音助理月活跃3880万排在中国第一，这种智能交互人工智能系统用的人越多，系统越好。今天随着我们上千名人工智能方面的工程师不断改进，用的人越来越多，我们人工智能产品的实用性就越来越好。<br/>日经中文网统计了一下，小米在AI专利申请量全世界排在11位，这个数字表明了小米在人工智能领域的投入远超大家想象。<br/>今天几乎大部分人用的是全面屏手机，这个概念是由小米发明的，小米拥有这个领域100多项专利，我们在这个方面做了大量的预研工作，带动了全行业的换代和升级。<br/>5、6年前小米开始做手机芯片，手机芯片需要很长时间，两年前小米发布了第一代芯片。<br/>在5G研发方面，今年2月份小米在巴塞罗那世界移动通信大会 （简称：MWC）上发布了第一部5G手机，所以小米在研发上下了很多功夫投入很大。<br/>2、设计。小米MIX手机这一款现代消费电子产品，先后被芬兰国家艺术馆、法国蓬皮杜艺术中心和德国慕尼黑国际设计博物馆收藏，这表达了小米在产品上设计水平的能力。我们横扫了这个星球上所有最高级别的设计奖，包括IDEA设计金奖、红点最佳设计奖、IF设计金奖和日本优良设计金奖。搞不好我们是中国公司里面拿设计奖最多的公司，总计拿了200多项。。<br/>3、品质。我在内部讲的一句话，后来成了金句：“创新决定我们能飞得有多高，品质决定我们能够走多远。”我自己亲自参加公司内部很多次质量大会，我在那些会上引用总理讲的话“质量之魂存于匠心”。过去一年多，凡是中国有质量奖评选的活动我们几乎都参与了，而且基本上我们都获奖了。<br/>第三，高效率的商业模型。<br/>中国的产品为什么做不好还有最重要一点是效率。各个环节消费很低，我们做一个好产品以后成本很高，所以需要卖的非常贵才能挣钱。提高效率以后，我可以卖的非常便宜，但是把钱主要用在研发上，这样可以很便宜价钱买到其实质量非常好的产品。谈到效率这个问题，我再科普一下商品价格构成。<br/>一个商品的价格包括原材料和制造成本、研发设计分摊成本、市场推广及广告、销售及渠道成本和利润。你们买到任何一个商品最核心是两块：原材料及制造研发、设计成本。我们的关键点是怎么来降低和消解这五项核心成本。我们要把这个产品做好，原材料及制造研发、设计成本这两项省不掉的，甚至你可能要做的更夸张。我在创办小米的时候，10来个人，一千万人民币的公司，初期选择供应商，不是苹果供应商不用，不是三星旗舰手机供应商不用，门槛巨高无比。为什么小米手机在初期一出厂的时候就是世界顶级产品和品质？初期小米出厂的时候，平均售价是1999元人民币，当时旗舰手机都卖4千到7千的时候，我们2千块钱手机的成本几乎都用在生产和制造上。<br/>研发分摊成本是什么呢？它不是研发成本总量，是研发分摊到每一个东西上的成本，是研发做一款产品的成本。我刚开始做手机的时候，同行每年做100款手机，每一款手机花1000万人民币，花了10亿人民币研发，你们会觉得它很厉害，实际上我花了10倍，我只做一款手机，花10倍就是1亿人民币，我总的研发成本就是它的1/10，但是我花在一款产品上的钱是它的10倍。结果我这款手机拿上来，小米第一代卖了760万台。我们做很少的SKU，投的研发的钱却是很多的。<br/>我办小米的时候又做了一个大胆的尝试，前五年零广告费，一分钱没有花过。关于销售和渠道成本，我又做了一个很夸张的动作，就是前店后厂。左边生产了一部手机，右边我开一个网站叫小米网直接卖了。<br/>接着问题是什么呢？我们只挣一两个点，我们怎么能够找到这个星球上最优秀的人，我们怎么能够有足够研发投资，甚至有足够外部投资帮助我们渡过难关。这个事情很难吧，因为利润很少，你很容易经营不好就破产了。我提出来“铁人三项”，实际上我们既做硬件，也做软件，也做互联网，也做服务，我们得垂直整合，在你看不到，在你没有意识到的地方我们也要认真经营，所以我们采用了互联网模式来做实体经济。<br/>我研究过Costco的利润，这么低的毛利基本上不可能挣钱，最高毛利不超过14%，综合毛利在7%到8%，它居然不亏钱，我就琢磨怎么挣钱，第一个挣钱，它开始收会员费，后来开始推广Costco信用卡，因为美国信用卡扣2到3个点，所以它实际上是早期互联网+的模型。<br/>我们去年互联网服务收入到了160亿，同比增长61.2%，这基本上就是我们的利润来源，前年是99亿，去年160亿，互联网服务的增长还是非常快。<br/>总结下我的观点：性价比不是战术，性价比是一种价值观，性价比是一种商业模式，性价比是一种核心能力，背后是对效率的极致追求和死磕自己的决心。<br/>这里面含有两条重要的内涵：第一条，极致的效率；第二条，严于律己死磕自己。<br/>03、小米的未来：手机+AIoT<br/>对于小米，未来就是5G的春天，因为所有人的手机都要换5G。我更看好是5G+AI+IoT，我认为这是下一代超级互联网，能够把所有的东西连起来。小米五年多前开始做物联网，现在做到世界第一了，但是这个领域会继续加速。去年年底小米定了未来十年战略是手机+AIoT双引擎，我们承诺在未来五年里面投资100个亿在AIoT。<br/>我在过去9年在管理上也做了很多大胆的尝试。我们叫去title（职位）；去KPI（关键业绩指标）；去管理。我有这些做法是因为我管公司20年以后想试一下，有没有真正高效的管理。我发现管理是一个悖论，管理就是控制风险，意味着不出现最坏的结果不翻船，但是也不保证出现最好的结果。<br/>十年前我就在想，有没有比大家传统意义上理解的管理更好的管理？后来我想明白了，更好的管理就是不管理。不管理比管理难很多，核心是找一帮优秀的人，找一帮不需要管理的人，塑造共同的愿景和目标，给足够激励就行了。什么样的人不需要管理？要有强大责任心、强大的自我驱动力，强大的梦想和抱负，心甘情愿每天12点下班，不用老板看着。我们小米刚开始就是这样的，我们找了一群非常热忱的人来做，这个模式干了八年以后，取得了巨大成功之后，我们还是需要很冷静地看待，它的问题是你怎么保证2万3千人都这样，他们散布在全球80多个国家，所以我们怎么能够继承发扬小米创业时候的优良特点，结合今天传统通用流程管理，怎么相结合，同时要小心过度管理。<br/>所以，我们又在摸索小米在第二个阶段创业的时候管理模式，干了很多事情：<br/>1、聚焦技术创新。<br/>2、强化总部管理职能，设立组织部和参谋部。<br/>3、梳理业务，提拔年轻干部。<br/>我在内部讲过，其实流程和规范有一点像练剑，手中有剑，我们怎么练到手中无剑心中有剑，怎么升华到无剑无我的阶段，这是武功最高的境界。在初期的时候，我们公司也很少有会议，基本上不怎么开会，这样干了很多年，现在在一步一步导入管理，因为实验完了以后，管一两千人没有问题，到几万人的话，还是需要怎么结合传统管理经验，我们现在正在探索的过程当中。<br/>我在内部提了一个目标“未来十年，增长十倍”。手机+AIoT就是增长计划背后很重要的逻辑。', 'created_at': 1583415919000, 'retweet_count': 12, 'reply_count': 11, 'fav_count': 14, 'truncated': False, 'commentId': 0, 'retweet_status_id': 0, 'symbol_id': None, 'description': '<a href=""http://xueqiu.com/S/01810"" target=""_blank"">$小米集团-W(01810)$</a> 雷军清华演讲：小米9年奇迹背后 未来十年增长十倍2019-04-05 4月4日，应清华大学国家金融研究院院长、IMF前副总裁朱民之邀，雷军在清华大学进行了一次演讲，主题是“小米9年：创新、变革与未来”，主要介绍了小米的创立过程和发展模式，并与与会的清华师生、校友与社会各界人...', 'type': '2', 'source_link': 'http://xueqiu.com/about/mobile-xueqiu', 'edited_at': None, 'pic': '', 'user': {'subscribeable': False, 'remark': None, 'status': 1, 'common_count': 0, 'domain': None, 'description': '马明哲：我们积蓄了这么多年的能量，无论公司还是股票，现在的爆发只是刚刚开始呢，好戏在后面。（2016年股东大会）', 'type': '1', 'location': None, 'id': 3818775985, 'url': None, 'screen_name': '等待就是忍耐', 'following': False, 'profile': '/3818775985', 'verified': False, 'blocking': False, 'donate_count': 0, 'friends_count': 84, 'followers_count': 265, 'st_color': '1', 'follow_me': False, 'verified_description': None, 'province': '浙江', 'city': '杭州', 'gender': 'm', 'status_count': 934, 'step': 'three', 'blog_description': None, 'verified_type': 0, 'last_status_id': 143050445, 'stats': None, 'recommend': None, 'intro': None, 'stock_status_count': None, 'allow_all_stock': False, 'stocks_count': None, 'name': None, 'group_ids': None, 'name_pinyin': None, 'screenname_pinyin': None, 'block_status': 0, '_source': None, 'verified_realname': False, 'verified_infos': None, 'profile_image_url': 'community/20177/1501661898351-1501661898978.jpg,community/20177/1501661898351-1501661898978.jpg!180x180.png,community/20177/1501661898351-1501661898978.jpg!50x50.png,community/20177/1501661898351-1501661898978.jpg!30x30.png', 'photo_domain': '//xavatar.imedao.com/'}, 'retweeted_status': None, 'target': '/3818775985/143037703', 'fragment': None, 'blocked': False, 'blocking': False, 'topic_pic': None, 'topic_symbol': None, 'topic_title': None, 'topic_desc': None, 'donate_count': 0, 'donate_snowcoin': 0, 'view_count': 7919, 'offer': None, 'mark': 0, 'comment': None, 'expend': True, 'rqtype': None, 'rqid': 0, 'firstImg': None, 'card': None, 'favorited': False, 'favorited_created_at': None, 'topic_pic_thumbnail': None, 'topic_pic_headOrPad': None, 'topic_pic_thumbnail_small': None, 'canEdit': True, 'shortStatusMultiPic': False, 'timeBefore': '03-05 21:45', 'mark_desc': None, 'reward_count': 0, 'reward_user_count': 0, 'paid_mention': None, 'is_answer': False, 'is_refused': False, 'is_reward': True, '_source': None, 'pic_sizes': [], 'like_count': 12, 'liked': False, 'promotion_pic': None, 'promotion_url': None, 'promotion_id': 0, 'quote_cards': None, 'is_ss_multi_pic': False, 'excellent_comments': None, 'source': 'iPad', 'is_bonus': False}, 'target': '/4658260427/143045191', 'fragment': None, 'blocked': False, 'blocking': False, 'topic_pic': None, 'topic_symbol': None, 'topic_title': None, 'topic_desc': None, 'donate_count': 0, 'donate_snowcoin': 0, 'view_count': 0, 'offer': None, 'mark': 0, 'comment': None, 'expend': False, 'rqtype': None, 'rqid': 0, 'firstImg': None, 'card': None, 'favorited': False, 'favorited_created_at': None, 'topic_pic_thumbnail': None, 'topic_pic_headOrPad': None, 'topic_pic_thumbnail_small': None, 'canEdit': False, 'shortStatusMultiPic': False, 'timeBefore': '03-05 22:48', 'mark_desc': None, 'reward_count': 0, 'reward_user_count': 0, 'paid_mention': None, 'is_answer': False, 'is_refused': False, 'is_reward': True, '_source': 'sc_1:9:01810', 'pic_sizes': [], 'like_count': 0, 'liked': False, 'promotion_pic': None, 'promotion_url': None, 'promotion_id': 0, 'quote_cards': None, 'is_ss_multi_pic': False, 'excellent_comments': None, 'source': 'iPhone', 'is_bonus': False}"
143041451,7806065303,1583418512000,啥时候公布财报啊,"{'id': 143041451, 'user_id': 7806065303, 'title': '', 'text': '<a href=""http://xueqiu.com/S/01810"" target=""_blank"">$小米集团-W(<span class=\'highlight\'>01810</span>)$</a> 啥时候公布财报啊', 'created_at': 1583418512000, 'retweet_count': 0, 'reply_count': 0, 'fav_count': 0, 'truncated': False, 'commentId': 0, 'retweet_status_id': 0, 'symbol_id': None, 'description': "" $小米集团-W( <span class='highlight'>01810</span> )$ 啥时候公布财报啊"", 'type': '0', 'source_link': 'http://xueqiu.com/about/mobile-xueqiu', 'edited_at': None, 'pic': '', 'user': {'subscribeable': False, 'remark': None, 'status': 1, 'common_count': 0, 'domain': None, 'description': '', 'type': '1', 'location': None, 'id': 7806065303, 'url': None, 'screen_name': '苍窘之手', 'following': False, 'profile': '/7806065303', 'verified': False, 'blocking': False, 'donate_count': 0, 'friends_count': 95, 'followers_count': 4, 'st_color': '1', 'follow_me': False, 'verified_description': None, 'province': '浙江', 'city': '宁波', 'gender': 'n', 'status_count': 60, 'step': 'three', 'blog_description': None, 'verified_type': 0, 'last_status_id': 143041451, 'stats': None, 'recommend': None, 'intro': None, 'stock_status_count': None, 'allow_all_stock': False, 'stocks_count': None, 'name': None, 'group_ids': None, 'name_pinyin': None, 'screenname_pinyin': None, 'block_status': 0, '_source': None, 'verified_realname': False, 'verified_infos': None, 'profile_image_url': 'community/20151/1424577058885.jpg,community/20151/1424577058885.jpg!180x180.png,community/20151/1424577058885.jpg!50x50.png,community/20151/1424577058885.jpg!30x30.png', 'photo_domain': '//xavatar.imedao.com/'}, 'retweeted_status': None, 'target': '/7806065303/143041451', 'fragment': None, 'blocked': False, 'blocking': False, 'topic_pic': None, 'topic_symbol': None, 'topic_title': None, 'topic_desc': None, 'donate_count': 0, 'donate_snowcoin': 0, 'view_count': 0, 'offer': None, 'mark': 0, 'comment': None, 'expend': False, 'rqtype': None, 'rqid': 0, 'firstImg': None, 'card': None, 'favorited': False, 'favorited_created_at': None, 'topic_pic_thumbnail': None, 'topic_pic_headOrPad': None, 'topic_pic_thumbnail_small': None, 'canEdit': True, 'shortStatusMultiPic': False, 'timeBefore': '03-05 22:28', 'mark_desc': None, 'reward_count': 0, 'reward_user_count': 0, 'paid_mention': None, 'is_answer': False, 'is_refused': False, 'is_reward': True, '_source': 'sc_1:10:01810', 'pic_sizes': [], 'like_count': 0, 'liked': False, 'promotion_pic': None, 'promotion_url': None, 'promotion_id': 0, 'quote_cards': None, 'is_ss_multi_pic': False, 'excellent_comments': None, 'source': 'Android', 'is_bonus': False}"
```
